# Q37: Stable Diffusion Architecture
**Topic:** LLM | **Difficulty:** Medium | **Time:** 15 min
**Sheet:** GrindLLM50

## Question
Explain the architecture of stable-diffusion.

## Detailed Answer

### Overview
Stable Diffusion performs diffusion in a **latent space** rather than pixel space, making it much more efficient.

### Architecture Components
```
Text Prompt → CLIP Text Encoder → Text Embeddings
                                        ↓
Random Noise → U-Net (iterative denoising) → Denoised Latent → VAE Decoder → Image
                ↑ (cross-attention to text)
```

**1. VAE (Variational Autoencoder)**
- **Encoder**: Compresses 512×512 image → 64×64 latent (8× downsampling)
- **Decoder**: Reconstructs latent → pixel image
- Working in latent space = 64× less compute than pixel space

**2. U-Net (Denoising Model)**
- Core neural network that learns to denoise
- Encoder-decoder with skip connections (U-shape)
- Contains:
  - **ResNet blocks**: Convolutional processing
  - **Self-attention**: Spatial attention within the image
  - **Cross-attention**: Conditions on text embeddings
  - **Time embedding**: Injects timestep information
- Predicts noise ε (or velocity) to subtract from noisy latent

**3. Text Encoder (CLIP)**
- Pre-trained CLIP text encoder (77 tokens max)
- SDXL: CLIP-ViT-L + OpenCLIP-ViT-bigG (two text encoders)
- Text embeddings are injected via cross-attention in the U-Net

**4. Scheduler / Sampler**
- Controls the denoising schedule (noise levels per step)
- Examples: DDPM, DDIM, Euler, DPM-Solver
- DDIM: Deterministic, fewer steps needed

### Classifier-Free Guidance (CFG)
$$\hat{\epsilon} = \epsilon_{uncond} + w \cdot (\epsilon_{cond} - \epsilon_{uncond})$$
- w = guidance scale (typically 7-12)
- Higher w = more adherence to text prompt, less diversity

### Evolution
| Version | Key Changes |
|---------|------------|
| SD 1.5 | 860M U-Net, CLIP text encoder |
| SDXL | Larger U-Net, dual text encoder, refiner |
| SD 3 / Flux | DiT (Transformer replaces U-Net), T5 text encoder |

## Key Takeaways
- Stable Diffusion = **VAE** (compress to latent) + **U-Net** (denoise) + **CLIP** (text encoding)
- Key innovation: diffusion in **latent space** = 64× more efficient than pixel space
- Text conditioning via **cross-attention** in the U-Net
- Latest models (SD3, Flux) replace U-Net with **DiT** (Diffusion Transformer)
