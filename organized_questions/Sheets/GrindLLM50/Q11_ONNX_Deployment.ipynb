## Question
Assume you have 2 models - one trained in (a) PyTorch, (b) TensorFlow. How will you convert this to ONNX and prepare it for deployment?

## Detailed Answer

### What is ONNX?
**Open Neural Network Exchange** — framework-agnostic model format.
- Standardized operators and graph representation
- Run models from any framework on any ONNX runtime

### (a) PyTorch → ONNX
```python
import torch

model = MyModel()
model.eval()
dummy_input = torch.randn(1, 3, 224, 224)

torch.onnx.export(
    model, dummy_input, 'model.onnx',
    input_names=['input'],
    output_names=['output'],
    dynamic_axes={'input': {0: 'batch'}, 'output': {0: 'batch'}},
    opset_version=17
)
```

### (b) TensorFlow → ONNX
```python
# Using tf2onnx
import tf2onnx
import tensorflow as tf

model = tf.keras.models.load_model('model.h5')
spec = (tf.TensorSpec((None, 224, 224, 3), tf.float32, name='input'),)
model_proto, _ = tf2onnx.convert.from_keras(model, input_signature=spec, opset=17)
```

### Preparing for Deployment
1. **Validate**: Compare ONNX output vs original model output
2. **Optimize**: `onnxruntime` graph optimizations (constant folding, fusion)
3. **Quantize**: INT8 quantization for faster inference
4. **Profile**: Benchmark latency and throughput
5. **Serve**: ONNX Runtime, Triton Inference Server, or TensorRT

### Benefits of ONNX
- **Cross-framework**: Train in PyTorch, deploy with TensorRT
- **Hardware-optimized**: ONNX Runtime has optimized kernels for CPU/GPU/NPU
- **Graph optimizations**: Operator fusion, constant folding
- **Standardized**: Single format for all deployment targets

## Key Takeaways
- PyTorch: use `torch.onnx.export()` with dynamic axes for batching
- TensorFlow: use `tf2onnx.convert.from_keras()` or `from_saved_model()`
- Always **validate numerical equivalence** after conversion
- ONNX Runtime provides 2-5x speedup over native PyTorch inference
