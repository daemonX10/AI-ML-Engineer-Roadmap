## Question
What distinguishes GPT from BERT and T5?

## Detailed Answer

### Architecture Comparison
| Feature | GPT | BERT | T5 |
|---------|-----|------|----|
| **Type** | Decoder-only | Encoder-only | Encoder-Decoder |
| **Attention** | Causal (unidirectional) | Bidirectional | Bidirectional enc, Causal dec |
| **Pre-training** | Next token prediction | MLM + NSP | Span corruption (denoising) |
| **Generation** | Yes (natural) | No (not designed for) | Yes |
| **Understanding** | Good | Excellent | Excellent |

### GPT (Generative Pre-trained Transformer)
- **Objective**: Predict next token $P(x_t | x_{<t})$
- **Strengths**: Text generation, in-context learning, few-shot
- **Weakness**: Unidirectional context limits understanding
- **Models**: GPT-2, GPT-3, GPT-4, LLaMA, Mistral

### BERT (Bidirectional Encoder Representations)
- **Objective**: Predict masked tokens using FULL context (forward + backward)
- **Strengths**: Understanding tasks (classification, NER, QA)
- **Weakness**: Cannot generate text naturally
- **Models**: BERT, RoBERTa, DeBERTa, ALBERT

### T5 (Text-to-Text Transfer Transformer)
- **Objective**: Everything as text-to-text ("translate English to French: ...")
- **Strengths**: Versatile, good at seq2seq tasks
- **Unique**: Frames ALL tasks as text generation
- **Models**: T5, mT5, Flan-T5, UL2

### Key Insight
The fundamental difference is the **attention mask**:
- GPT: See only past → good at generation
- BERT: See everything → good at understanding
- T5: Encode everything, generate with past only → good at both

### Current Trend
**Decoder-only models (GPT-style) have become dominant** because:
- Scale better with more data and compute
- In-context learning eliminates need for task-specific fine-tuning
- Can handle both understanding and generation

## Key Takeaways
- **GPT** = decoder-only, causal attention, excels at generation
- **BERT** = encoder-only, bidirectional, excels at understanding
- **T5** = encoder-decoder, text-to-text, versatile for seq2seq
- Modern trend: **decoder-only** (GPT-style) wins by scaling + in-context learning
