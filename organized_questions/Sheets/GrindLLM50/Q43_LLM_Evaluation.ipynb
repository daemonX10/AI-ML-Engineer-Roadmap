## Question
How can you evaluate the performance of Language Models?

## Detailed Answer

### Intrinsic Metrics

**1. Perplexity (PPL)**
$$\text{PPL} = \exp\left(-\frac{1}{T}\sum_{t=1}^{T} \log P(x_t|x_{<t})\right)$$
- Lower = better (1 = perfect prediction)
- Measures how "surprised" the model is by the text
- Only for autoregressive models

### Task-Specific Metrics

**2. Text Generation Quality**
| Metric | What It Measures |
|--------|------------------|
| BLEU | N-gram precision (translation) |
| ROUGE | N-gram recall (summarization) |
| METEOR | Precision + recall + alignment |
| BERTScore | Semantic similarity via embeddings |
| ChrF | Character-level F-score |

**3. Question Answering**
- Exact Match (EM): Binary correctness
- F1 Score: Token-level overlap

**4. Classification**: Accuracy, F1, AUC

### Benchmark Suites
| Benchmark | What It Tests |
|-----------|---------------|
| MMLU | Multi-task knowledge (57 subjects) |
| HumanEval | Code generation |
| GSM8K | Grade school math |
| TruthfulQA | Truthfulness |
| HellaSwag | Commonsense reasoning |
| ARC | Science questions |
| GPQA | PhD-level questions |

### Human Evaluation
- **Chatbot Arena**: ELO ratings from pairwise human comparisons
- **Side-by-side comparison**: A/B testing of model outputs
- **Likert scale**: Rate quality, helpfulness, harmlessness (1-5)

### LLM-as-Judge
- Use a strong LLM (GPT-4, Claude) to evaluate outputs
- Cost-effective alternative to human evaluation
- Correlates well with human preferences (~85%)

## Key Takeaways
- **Perplexity** for language modeling, **BLEU/ROUGE** for generation tasks
- **Benchmark suites** (MMLU, HumanEval, GSM8K) for comprehensive evaluation
- **Human evaluation** (Chatbot Arena) remains the gold standard
- **LLM-as-Judge** is a cost-effective proxy for human evaluation
