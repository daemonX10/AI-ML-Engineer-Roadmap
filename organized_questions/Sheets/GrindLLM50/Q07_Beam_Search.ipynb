## Question
Why is Beam Search preferred in predicting the output of a language model?

## Detailed Answer

### The Problem with Greedy Decoding
Greedy: Always pick the highest-probability token at each step.
- Can miss globally better sequences
- Example: "The cat sat on the" → greedy picks "mat" but "comfortable old sofa" might be better overall

### Beam Search
Keep **B** (beam width) best partial sequences at each step.

```
Step 1: Keep top-B first tokens
Step 2: For each beam, extend with all vocab → B×V candidates → keep top-B
Step 3: Repeat until EOS or max length
```

Score: $\log P(y) = \sum_{t=1}^{T} \log P(y_t | y_{<t})$

**Length normalization** (prevents bias toward shorter sequences):
$$\text{score}(y) = \frac{1}{T^\alpha} \sum_{t=1}^{T} \log P(y_t | y_{<t})$$

### When Beam Search Works Best
- **Translation**: Deterministic output, single correct answer → B=4-6
- **Summarization**: Factual, focused output
- **Speech recognition**: Constrained vocabulary

### When NOT to Use Beam Search
- **Open-ended generation**: Produces repetitive, boring text
- **Chatbots**: Need diversity and creativity
- **Storytelling**: Want surprising, varied outputs

→ Use **sampling** (temperature, top-k, top-p) instead for creative tasks.

### Comparison
| Method | Quality | Diversity | Speed |
|--------|---------|-----------|-------|
| Greedy | Low | None | Fast |
| Beam (B=4) | High | Low | Medium |
| Sampling | Medium | High | Fast |
| Top-p + temp | High | High | Fast |

## Key Takeaways
- Beam search = **approximate search** over sequence space with B candidates
- Best for **deterministic** outputs (translation, ASR, summarization)
- **Avoid for open-ended generation** — produces repetitive text
- Modern chatbots use **sampling** (top-p/temperature) instead of beam search
