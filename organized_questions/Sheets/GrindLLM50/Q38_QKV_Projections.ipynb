## Question
What are the QKV matrices and why do we use projections of it?

## Detailed Answer

### What are Q, K, V?
Given input embeddings $X \in \mathbb{R}^{n \times d}$:
- $Q = XW_Q$ — **Query**: "What am I looking for?"
- $K = XW_K$ — **Key**: "What do I contain?" (index)
- $V = XW_V$ — **Value**: "What information do I carry?"

### The Attention Mechanism
$$\text{Attention}(Q, K, V) = \text{softmax}\left(\frac{QK^T}{\sqrt{d_k}}\right)V$$

1. $QK^T$: Each query asks "how relevant is each key to me?" → attention scores
2. Softmax: Normalize scores → attention weights (probabilities)
3. Weighted sum of V: Aggregate values based on relevance

### Why Projections? (Why not use X directly?)

**1. Separate Roles**
Without projections: $\text{softmax}(XX^T)X$
- Q, K, V are all the same thing — very limited expressiveness
- Projections let Q, K, V learn **different representations** for different roles

**2. Dimension Flexibility**
- $W_Q, W_K \in \mathbb{R}^{d \times d_k}$ — can project to different dimensions
- $d_k$ can differ from $d_{model}$ (needed for multi-head: $d_k = d_{model}/h$)

**3. Learned Compatibility**
- The model learns WHAT to match (Q) and WHAT to match against (K)
- These are different concepts! E.g., a verb (Q) might look for its subject (K)
- Raw embeddings wouldn't specialize for this matching

**4. Multi-Head Diversity**
- Each head has its own $W_Q^i, W_K^i, W_V^i$
- Different projections → heads attend to different relationships
- Without projections, all heads would compute the same thing

### Analogy
- **Library**: Q = your search query, K = book index card, V = book content
- You don't compare the book itself to your query — you compare your query to the index
- Then retrieve the relevant book's content

## Key Takeaways
- Q, K, V are **learned projections** of the same input, each serving a different role
- **Without projections**: self-attention reduces to $\text{softmax}(XX^T)X$ — too limited
- Projections enable **separate matching** (QK) and **information retrieval** (V)
- Critical for multi-head attention: each head learns different Q/K/V projections
