## Question
At a high level, what is RLHF? Why is this a novel paradigm compared to self-supervised learning?

## Detailed Answer

### RLHF Pipeline
```
Step 1: Pre-train LLM (self-supervised on text)
Step 2: Supervised Fine-Tuning (SFT) on high-quality demonstrations
Step 3: Train Reward Model (RM) on human preference pairs
Step 4: Optimize LLM via RL (PPO) using reward model
```

### Step 3: Reward Model Training
- Human annotators rank model outputs: A > B
- Train RM to predict which output humans prefer:
$$L_{RM} = -\log \sigma(r(x, y_w) - r(x, y_l))$$
where $y_w$ = preferred output, $y_l$ = rejected output

### Step 4: RL Optimization (PPO)
$$\max_\theta \mathbb{E}_{x \sim D, y \sim \pi_\theta}[r(x, y)] - \beta D_{KL}(\pi_\theta \| \pi_{ref})$$
- Maximize reward while staying close to reference model (prevents reward hacking)
- KL penalty prevents model from gaming the reward model

### Why Novel vs Self-Supervised?
| Aspect | Self-Supervised | RLHF |
|--------|----------------|------|
| Signal | Next token prediction | Human preference |
| Objective | Predict text distribution | Be helpful, harmless, honest |
| What it captures | Factual knowledge, grammar | Values, safety, style |
| Training data | Raw text | Human comparisons |
| Optimization | MLE | RL (PPO or DPO) |

Self-supervised learning optimizes for **plausibility** (what text is likely).
RLHF optimizes for **preference** (what humans actually want).

### DPO (Direct Preference Optimization) — Simpler Alternative
$$L_{DPO} = -\log \sigma\left(\beta \log \frac{\pi_\theta(y_w|x)}{\pi_{ref}(y_w|x)} - \beta \log \frac{\pi_\theta(y_l|x)}{\pi_{ref}(y_l|x)}\right)$$
- No reward model needed — directly optimize policy from preferences
- Simpler, more stable than PPO
- Used in LLaMA 2, Zephyr, many open-source models

## Key Takeaways
- RLHF bridges the gap between **predicting text** and **being helpful**
- The reward model captures human preferences that can't be expressed in simple loss functions
- **DPO** is replacing PPO-based RLHF — simpler and equally effective
- The key insight: MLE trains models to mimic text, RLHF trains them to satisfy humans
