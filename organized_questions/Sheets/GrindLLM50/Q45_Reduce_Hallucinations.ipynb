# Q45: Reducing Hallucinations in LLMs

## Question
How do we reduce hallucinations in a model?

## Detailed Answer

### Types of Hallucinations
- **Factual**: Generating false facts ("The Eiffel Tower is in London")
- **Faithful**: Output contradicts the provided context
- **Intrinsic**: Contradicts itself within the same response

### Mitigation Strategies

**1. Retrieval-Augmented Generation (RAG)**
- Ground responses in retrieved documents
- Model cites sources → verifiable claims
- Best approach for factual accuracy

**2. Better Training
- **RLHF**: Reward model penalizes hallucinated content
- **DPO**: Direct preference optimization with hallucination-aware pairs
- Train on high-quality, factual data
- Include "I don't know" examples in training data

**3. Decoding Strategies**
- Lower temperature (more deterministic)
- Factual nucleus sampling (constrain to high-probability tokens)
- Self-consistency: Sample multiple chain-of-thoughts, take majority answer

**4. Chain-of-Thought / Reasoning**
- Force model to show reasoning steps
- Errors in reasoning are easier to detect
- Self-reflection: "Verify your answer step by step"

**5. Post-hoc Verification**
- Use a separate model to fact-check generated claims
- Entailment checking: Does the context support this claim?
- Confidence calibration: Flag low-confidence outputs

**6. Prompt Engineering**
- "Only answer based on the provided context"
- "If you're not sure, say 'I don't know'"
- Provide relevant context in the prompt

**7. Architecture / Model-Level**
- Attribution heads: Model learns to cite sources
- Constrained decoding: Only generate tokens supported by evidence
- Knowledge grounding: Connect model to knowledge graph

### Effectiveness Ranking
1. **RAG** (most effective for factual tasks)
2. **RLHF with hallucination-aware training**
3. **Chain-of-thought + self-verification**
4. **Prompt engineering** (quick win, no training)
5. **Temperature/sampling control**

## Key Takeaways
- **RAG** is the most effective approach — ground in retrieved evidence
- **RLHF/DPO** can train models to say "I don't know" instead of hallucinating
- **Chain-of-thought + self-verification** catches reasoning errors
- No single approach eliminates hallucinations — use multiple strategies together
