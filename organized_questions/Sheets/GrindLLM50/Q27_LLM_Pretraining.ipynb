## Question
How are LLMs pre-trained?

## Detailed Answer

### Pre-training Objective
**Causal Language Modeling (CLM)**: Predict the next token given all previous tokens.
$$\mathcal{L} = -\sum_{t=1}^{T} \log P(x_t | x_{<t}; \theta)$$

### Pipeline

**1. Data Collection**
- Web crawl (CommonCrawl, C4, RefinedWeb, FineWeb)
- Books, code (GitHub, StackOverflow), Wikipedia, academic papers
- Trillions of tokens (LLaMA 3: 15T tokens)

**2. Data Processing**
- Deduplication (exact + fuzzy hash-based)
- Quality filtering (perplexity, heuristic rules, classifier-based)
- Toxic content removal
- Language identification and balancing

**3. Tokenization**
- BPE (Byte-Pair Encoding) or SentencePiece
- Vocabulary: 32K-128K tokens

**4. Training**
- Optimizer: AdamW (β₁=0.9, β₂=0.95)
- Learning rate: Cosine schedule with warmup
- Batch size: Millions of tokens (gradually increased)
- Sequence length: 2048-8192 tokens
- Duration: Weeks to months on thousands of GPUs

**5. Distributed Training**
- Data parallelism: Same model, different data batches
- Tensor parallelism: Split layers across GPUs
- Pipeline parallelism: Split model layers
- FSDP or DeepSpeed ZeRO for memory efficiency

### Compute Requirements
| Model | Training Compute | GPUs |
|-------|-----------------|------|
| GPT-3 (175B) | ~3640 PF-days | ~1000 A100 |
| LLaMA-7B | ~184K GPU-hours | 2048 A100 |
| LLaMA-70B | ~1.7M GPU-hours | 2048 A100 |

## Key Takeaways
- Pre-training objective: **next token prediction** (causal LM)
- Data quality > data quantity — filtering and dedup are critical
- Training uses trillions of tokens over weeks on thousands of GPUs
- Modern LLMs: BPE tokenizer → Transformer decoder → cosine LR → AdamW
