# Q40: Transformers vs LSTMs

## Question
How are transformer-based architectures better than LSTM-based architectures in NLP tasks?

## Detailed Answer

### Key Differences
| Aspect | LSTM | Transformer |
|--------|------|-------------|
| **Parallelization** | Sequential (step-by-step) | Fully parallel |
| **Long-range deps** | Degrades over distance | Constant (O(1) path) |
| **Training speed** | Slow (can't parallelize) | Fast (GPU efficient) |
| **Memory** | Compressed hidden state | Explicit attention to all tokens |
| **Context window** | Decays exponentially | Full context (with FlashAttention) |

### Why Transformers Win

**1. Parallelization**
- LSTM: Must process tokens sequentially: $h_t = f(h_{t-1}, x_t)$
- Transformer: All positions computed simultaneously → massive GPU parallelism
- 10-100x faster training

**2. Long-Range Dependencies**
- LSTM: Information must pass through every step → vanishing/exploding gradients
- Even with gates, long-range dependencies degrade (~100-500 tokens practical)
- Transformer: Direct attention between ANY two positions → O(1) path length

**3. Scaling**
- Transformers scale predictably with data and compute (scaling laws)
- LSTMs don't benefit as much from scale
- All modern language models (1B+ params) are Transformer-based

**4. Pre-training**
- BERT/GPT-style pre-training works naturally with Transformers
- LSTMs don't benefit as much from self-supervised pre-training at scale

### When LSTMs Might Still Be Used
- Extremely resource-constrained edge devices
- Real-time streaming (constant memory per step)
- Simple time-series tasks with short sequences

### Modern Alternatives to Both
- **Mamba/S4**: State-space models — linear complexity, recurrent-like but parallelizable
- **RWKV**: Combines RNN efficiency with Transformer-like training

## Key Takeaways
- Transformers are better due to **parallelization** and **direct long-range connections**
- LSTMs are sequential → can't scale on modern GPU hardware
- Transformer scaling laws enable predictable performance improvement with more compute
- **Mamba/SSM** models combine best of both: linear complexity + parallelizable training
