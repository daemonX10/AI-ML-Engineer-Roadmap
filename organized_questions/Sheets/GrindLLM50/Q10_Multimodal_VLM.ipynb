## Question
How do multi-modal vision-language models like Gemma 3 or LLaVA integrate visual and textual information? Describe the architecture, training process, and major innovations.

## Detailed Answer

### General Architecture
```
Image → Vision Encoder → Adapter/Projector → LLM ← Text Tokens
                                               ↓
                                           Response
```

### Components

**1. Vision Encoder**
- Typically a pre-trained ViT (Vision Transformer): CLIP ViT, SigLIP, DINOv2
- Produces patch-level visual features: image → grid of embeddings

**2. Adapter/Projector**
- Bridges the modality gap between vision and language embeddings
- LLaVA: Simple linear projection (2-layer MLP)
- Gemma 3: SigLIP encoder + linear projection
- Cross-attention adapters (Flamingo): More complex but richer integration

**3. LLM Backbone**
- Receives projected visual tokens + text tokens as a single sequence
- Processes them with standard self-attention
- Visual tokens are treated as a "visual prefix"

### Training Process

**Stage 1: Pre-training (Alignment)**
- Train only the projector on image-text pairs
- Keep vision encoder and LLM frozen
- Dataset: Large-scale image-caption pairs

**Stage 2: Instruction Tuning**
- Fine-tune projector + LLM (optionally vision encoder too)
- Dataset: Visual question answering, image captioning, visual reasoning
- GPT-4 generated multimodal instruction data (LLaVA approach)

### Key Innovations
| Model | Innovation |
|-------|------------|
| LLaVA | Simple MLP projector, visual instruction tuning |
| Gemma 3 | SigLIP encoder, pan-and-scan for high-res |
| Flamingo | Cross-attention gated layers |
| GPT-4V | Native multimodal training (not disclosed) |
| Florence-2 | Unified visual grounding + captioning |

## Key Takeaways
- VLMs = Vision Encoder + Adapter + LLM backbone
- **Two-stage training**: alignment (projector only) → instruction tuning (LLM + projector)
- A simple **MLP projector** (LLaVA) works surprisingly well
- Visual tokens join the text sequence — the LLM processes them with standard attention
