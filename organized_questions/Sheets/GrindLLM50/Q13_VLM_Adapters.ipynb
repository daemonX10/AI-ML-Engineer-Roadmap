## Question
Why do VLMs like LLaVA or Florence-VL use adapters / cross-attention instead of directly feeding visual embeddings into LLMs?

## Detailed Answer

### The Modality Gap
Vision encoder embeddings live in a **different representation space** than LLM text embeddings.
- CLIP ViT output: 768d or 1024d, trained with contrastive loss
- LLM embedding: 4096d (LLaMA), trained for next-token prediction
- Directly feeding visual features would be like speaking French to someone who only understands English

### Why Adapters?

**1. Dimensionality mismatch**: Vision encoder output ≠ LLM input dimension
- Need projection: $\mathbb{R}^{d_{vision}} \rightarrow \mathbb{R}^{d_{LLM}}$

**2. Semantic alignment**: Even after projection, raw visual features may not be
"meaningful" to the LLM. The adapter learns to translate visual concepts into
the language model's "vocabulary" of representations.

**3. Training efficiency**: Keep both pre-trained models frozen, only train the adapter
- LLaVA Stage 1: Only train MLP projector (~50M params vs 7B LLM)
- Much faster, less data needed

**4. Preserve pre-trained knowledge**: Fine-tuning the LLM directly on visual tasks
risks catastrophic forgetting of language abilities.

### Types of Adapters
| Adapter Type | Model | How |
|-------------|-------|-----|
| **Linear/MLP** | LLaVA | Simple 2-layer MLP projection |
| **Cross-attention** | Flamingo | Interleave cross-attention layers in LLM |
| **Q-Former** | BLIP-2 | Learnable query tokens attend to visual features |
| **Perceiver** | General | Fixed-size bottleneck attention |

Simpler adapters (LLaVA's MLP) are surprisingly effective — the key insight is
that the LLM is powerful enough to understand visual information once it's in
the right representation space.

## Key Takeaways
- Vision and language models have **different representation spaces** — adapters bridge the gap
- A simple **MLP projector** (LLaVA) is surprisingly effective
- Adapters enable **training efficiency** — don't need to fine-tune the full LLM
- Complex adapters (Q-Former, cross-attention) can be better but add training complexity
