# Q29: Shrinking Transformer Models
**Topic:** LLM | **Difficulty:** Medium | **Time:** 10 min
**Sheet:** GrindLLM50

---

## Question
Transformer models are expensive. What are ways to shrink them?

## Detailed Answer

### 1. Quantization
Reduce numerical precision of weights/activations.
| Method | Precision | Compression | Quality |
|--------|-----------|------------|--------|
| FP16/BF16 | 16-bit | 2x | ~Lossless |
| INT8 (LLM.int8) | 8-bit | 4x | Good |
| INT4 (GPTQ, AWQ) | 4-bit | 8x | Good |
| GGUF 2-4 bit | Mixed | 8-16x | Acceptable |

### 2. Knowledge Distillation
Train a smaller **student** model to mimic a larger **teacher**.
- DistilBERT: 6-layer student from 12-layer BERT (40% smaller, 97% performance)
- Student learns from teacher's soft probabilities (dark knowledge)
$L = \alpha L_{CE} + (1-\alpha) T^2 D_{KL}(\sigma(z_t/T) \| \sigma(z_s/T))$

### 3. Pruning
Remove redundant components.
- **Unstructured**: Zero out individual weights (sparse matrices)
- **Structured**: Remove entire heads, layers, or neurons
- Head pruning: Some attention heads contribute minimally
- Layer dropping: Remove intermediate layers

### 4. Architecture Efficiency
- **MoE**: More capacity with less active compute
- **GQA/MQA**: Fewer KV heads → less memory
- **Sliding window**: O(n·w) instead of O(n²)
- **Mamba/SSM**: Replace attention entirely

### 5. Other Techniques
- **Weight sharing**: Tie embedding layers
- **Low-rank factorization**: Decompose weight matrices
- **Dynamic inference**: Early exit for easy inputs
- **Vocabulary pruning**: Reduce tokenizer vocabulary

## Key Takeaways
- **Quantization** (INT4/INT8) is the easiest and most impactful compression
- **Distillation** creates smaller models (DistilBERT = 40% smaller, 97% quality)
- **Pruning** can remove 30-50% of attention heads with minimal quality loss
- In practice: quantize first, then consider distillation if more compression needed
