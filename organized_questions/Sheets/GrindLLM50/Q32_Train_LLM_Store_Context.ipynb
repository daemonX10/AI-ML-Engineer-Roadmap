# Q32: Training an LLM to Store Specific Context
**Topic:** LLM | **Difficulty:** Medium | **Time:** 10 min
**Sheet:** GrindLLM50

---

## Question
Is there a way to train an LLM to store a specific context?

## Detailed Answer

### Methods to Store Specific Context

**1. Fine-tuning on Specific Data**
- Train on documents containing the context
- Model memorizes facts in its weights
- Risk: Overfitting to specific phrasing, catastrophic forgetting

**2. RAG (Retrieval-Augmented Generation)**
- Store context in an external vector database
- Retrieve relevant chunks at inference time
- **Best approach** for dynamic, factual knowledge
- Context lives in the DB, not in model weights

**3. Prompt/Context Window**
- Simply provide context in the prompt (system message or user message)
- Works for small amounts of context (<128K tokens)
- No training needed

**4. Knowledge Distillation**
- Train a model specifically on QA pairs derived from the context
- The context becomes "baked in" to the model's parameters

**5. Continued Pre-training**
- Continue pre-training (next-token prediction) on specific documents
- More stable than instruction fine-tuning for knowledge injection
- Need to mix with general data to prevent forgetting

### Best Practices
| Need | Best Approach |
|------|---------------|
| Dynamic knowledge | RAG |
| Static domain knowledge | Continued pre-training + LoRA |
| Small context (<128K) | Put in prompt |
| Specific behavior + knowledge | Fine-tune + RAG |

## Key Takeaways
- **RAG** is the best approach for storing retrievable context
- Fine-tuning embeds knowledge in weights but risks forgetting
- For small context, just **put it in the prompt** (simplest solution)
- Continued pre-training is more stable than instruction fine-tuning for knowledge injection
