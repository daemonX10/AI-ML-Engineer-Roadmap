## Question
How many parameters does BERT have? How does DistilBERT reduce this?

## Detailed Answer

### BERT Parameter Count

**BERT-base**: 110M parameters
- 12 Transformer layers, 768 hidden size, 12 attention heads
- Vocabulary: 30,522 tokens (WordPiece)

**BERT-large**: 340M parameters
- 24 layers, 1024 hidden size, 16 attention heads

**Parameter Breakdown (BERT-base):**
| Component | Params |
|-----------|--------|
| Token embeddings (30522 × 768) | 23.4M |
| Position embeddings (512 × 768) | 0.4M |
| Segment embeddings (2 × 768) | ~0M |
| Transformer layers (12×) | ~85M |
| Pooler | ~0.6M |
| **Total** | **~110M** |

### DistilBERT: Knowledge Distillation

**Size**: 66M parameters (40% reduction)
**Architecture**: 6 layers (vs 12), same hidden size 768, 12 heads
**Performance**: 97% of BERT's performance, 60% faster

### How Distillation Works
**Triple loss:**
$$L = \alpha L_{CE} + \beta L_{MLM} + \gamma L_{cos}$$

1. **Distillation loss** ($L_{CE}$): KL divergence between teacher (BERT) and student (DistilBERT) soft probabilities
$$L_{CE} = \sum_i t_i \log \frac{t_i}{s_i}$$
where $t_i = \text{softmax}(z_{teacher}/T)$ at temperature $T$

2. **MLM loss** ($L_{MLM}$): Standard masked LM loss on ground truth

3. **Cosine loss** ($L_{cos}$): Align student hidden states to teacher hidden states

### Key Design Choices
- **Remove every other layer** (layers 0,2,4,6,8,10 → 6 layers)
- Keep hidden dimension the same (768)
- Initialize student from teacher weights
- Temperature T=8 for soft targets

## Key Takeaways
- BERT-base: **110M params** (12 layers, 768 hidden, 12 heads)
- DistilBERT: **66M params** (6 layers) → 40% smaller, 97% quality, 60% faster
- Distillation = train student on teacher's soft outputs (dark knowledge)
- Triple loss: KL divergence + MLM + cosine alignment
