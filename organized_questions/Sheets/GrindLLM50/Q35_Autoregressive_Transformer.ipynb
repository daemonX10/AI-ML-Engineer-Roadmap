# Q35: Is the Transformer Autoregressive?
**Topic:** LLM | **Difficulty:** Hard | **Time:** 15 min
**Sheet:** GrindLLM50

## Question
Is the transformer an auto-regressive model?

## Detailed Answer

### Short Answer
**It depends on the variant.** The Transformer architecture itself is not inherently autoregressive — it can be used in both autoregressive and non-autoregressive ways.

### Autoregressive Transformers (Yes)
**Decoder-only models** (GPT, LLaMA, Mistral) ARE autoregressive:
- Generate one token at a time
- Each token conditioned on all previous tokens
- $P(x) = \prod_{t=1}^{T} P(x_t | x_{<t})$
- Use causal masking to prevent future information leakage

**Training vs Inference:**
- **Training**: Process full sequence in parallel (teacher forcing)
  - Causal mask enables parallel computation despite autoregressive structure
  - All positions are computed simultaneously
- **Inference**: Sequential generation (one token at a time)
  - Must generate left-to-right
  - KV caching avoids redundant computation

### Non-Autoregressive Transformers (No)
**Encoder models** (BERT) are NOT autoregressive:
- Process all tokens simultaneously with bidirectional attention
- Don't generate tokens sequentially
- Used for understanding, not generation

**Non-Autoregressive Translation (NAT):**
- Generate all output tokens in parallel
- Faster but usually lower quality

### Diffusion LLMs (New Direction)
- Models like MDLM generate text via diffusion process
- Non-autoregressive: denoise all positions simultaneously
- Still experimental but growing area

### Summary
| Model Type | Autoregressive? | Example |
|-----------|----------------|--------|
| Decoder-only | Yes | GPT, LLaMA |
| Encoder-only | No | BERT |
| Encoder-Decoder | Decoder part yes | T5, BART |
| NAT | No | Parallel translation |

## Key Takeaways
- **Decoder-only** Transformers (GPT, LLaMA) ARE autoregressive
- **Encoder-only** (BERT) and NAT models are NOT autoregressive
- Autoregressive training uses **teacher forcing** (parallel) with causal mask
- Autoregressive inference is sequential (token by token)
