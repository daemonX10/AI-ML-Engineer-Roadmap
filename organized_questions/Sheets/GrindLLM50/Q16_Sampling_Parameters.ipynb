## Question
Discuss the impact of key sampling parameters for an LLM: (a) Temperature (b) top_k (c) top_p (d) freq and presence penalty (e) max_length

## Detailed Answer

### (a) Temperature
Scales logits before softmax: $p_i = \frac{e^{z_i / T}}{\sum_j e^{z_j / T}}$

| Temperature | Effect |
|-------------|--------|
| T → 0 | Greedy (argmax), deterministic |
| T = 1.0 | Model's trained distribution |
| T > 1.0 | Flatter distribution, more random |

- **Low T (0.1-0.3)**: Factual Q&A, code generation
- **Medium T (0.7)**: General chat
- **High T (1.0-1.5)**: Creative writing, brainstorming

### (b) Top-K Sampling
Keep only the K most probable tokens, redistribute probability mass.
- K=1 → greedy decoding
- K=50 → typical default
- **Problem**: Fixed K doesn't adapt. When distribution is peaked, K=50 includes many bad tokens. When flat, K=50 might exclude good ones.

### (c) Top-P (Nucleus) Sampling
Keep the smallest set of tokens whose cumulative probability ≥ P.
$$\text{Top-p set} = \{t_1, t_2, ...\} \text{ s.t. } \sum p(t_i) \geq p$$
- P=0.9 → keep tokens until 90% mass is covered
- **Adaptive**: Narrows/widens automatically based on distribution shape
- Preferred over top-k in practice

### (d) Frequency & Presence Penalty
Applied to logits to reduce repetition:

$$z_i' = z_i - \text{freq\_penalty} \times c_i - \text{presence\_penalty} \times \mathbb{1}[c_i > 0]$$

| Penalty | How | Effect |
|---------|-----|--------|
| Frequency | Proportional to count | Penalizes more each repetition |
| Presence | Binary (appeared or not) | Encourages new topics |

### (e) Max Length
- Hard cap on generation length
- Prevents runaway generation
- Affects cost (more tokens = more compute/money)
- Set based on task: summarization (200), code (2000), chat (1000)

## Key Takeaways
- **Temperature** controls randomness — low for factual, high for creative
- **Top-p** is preferred over top-k — adapts to distribution shape
- **Frequency penalty** reduces repetition proportionally, **presence penalty** encourages topic diversity
- In practice, combine: temperature=0.7 + top_p=0.9 + frequency_penalty=0.3
