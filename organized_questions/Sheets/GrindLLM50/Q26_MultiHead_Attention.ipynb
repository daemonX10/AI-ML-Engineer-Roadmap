## Question
Why do we need a Multi-Head Attention in Transformer-based architectures?

## Detailed Answer

### What is Multi-Head Attention?
Instead of one attention function, compute **h parallel attention heads**, each with its own projections:
$$\text{MultiHead}(Q,K,V) = \text{Concat}(\text{head}_1, ..., \text{head}_h) W^O$$
$$\text{head}_i = \text{Attention}(QW_i^Q, KW_i^K, VW_i^V)$$

### Why Multiple Heads?

**1. Different Attention Patterns**
- Head 1 might focus on syntactic relationships (subject-verb)
- Head 2 might capture coreference (pronouns → antecedents)
- Head 3 might attend to positional proximity
- Single head = single attention pattern per position

**2. Multiple Representation Subspaces**
- Each head projects to a different $d_k$-dimensional subspace
- Enables attending to different aspects of the representation simultaneously
- Like having h different "views" of the data

**3. Computational Efficiency**
- h heads with $d_k = d_{model}/h$ has **same total cost** as single head with $d_{model}$
- Parallel computation over heads (no sequential dependency)
- Each head is smaller → attention matrix is more expressive collectively

**4. Training Stability**
- Multiple heads provide redundancy
- If one head learns a bad pattern, others compensate
- More diverse gradient signals

### Observations
- Research shows different heads learn complementary patterns
- Some heads can be pruned without quality loss → not all equally important
- Typical: h=8 (BERT), h=32 (LLaMA-7B), h=64 (LLaMA-70B)

## Key Takeaways
- Multi-head = **parallel attention in different subspaces**
- Each head can capture different relationships (syntax, coreference, proximity)
- Same compute as single-head with $d_{model}$ attention — just split into h heads of $d_k = d_{model}/h$
- Critical for Transformer expressiveness — single head is too limited
