## Question
How does BERT training work?

## Detailed Answer

### Architecture
BERT = **Bidirectional Encoder Representations from Transformers**
- Encoder-only Transformer (no decoder)
- Processes all tokens simultaneously (bidirectional context)

### Pre-training Objectives

**1. Masked Language Modeling (MLM)**
- Randomly mask 15% of input tokens
- Of masked positions:
  - 80% → replaced with [MASK]
  - 10% → replaced with random token
  - 10% → kept unchanged
- Model predicts original token at masked positions
- Loss: Cross-entropy on masked tokens only

**2. Next Sentence Prediction (NSP)**
- Input: [CLS] Sentence A [SEP] Sentence B [SEP]
- Binary classification: Is B the actual next sentence after A?
- 50% true pairs, 50% random pairs
- Later shown to be less important (RoBERTa removed it)

### Input Representation
$$\text{Input} = \text{Token Embedding} + \text{Segment Embedding} + \text{Position Embedding}$$
- Token: WordPiece vocabulary (30K tokens)
- Segment: Which sentence (A or B)
- Position: Learned positional embeddings (max 512)

### Fine-tuning
Add a task-specific head on top:
| Task | Head | Where |
|------|------|-------|
| Classification | Linear([CLS]) | [CLS] output |
| Token classification (NER) | Linear(each token) | All token outputs |
| QA (SQuAD) | Start/End linear | All token outputs |

### BERT vs RoBERTa
RoBERTa improvements:
- Removed NSP objective
- Dynamic masking (different patterns each epoch)
- Larger batches and more data
- Longer training

## Key Takeaways
- BERT uses two objectives: **MLM** (predict masked tokens) + **NSP** (next sentence)
- **Bidirectional context** is the key advantage over GPT-style models
- Fine-tune by adding task-specific heads on [CLS] or token outputs
- RoBERTa showed that NSP isn't necessary and dynamic masking is better
