## Question
Why do transformers need Positional Encodings?

## Detailed Answer

### The Problem
Self-attention is **permutation-equivariant**: it computes the same output regardless of token order.
- If you shuffle the input tokens, the attention scores reshuffle identically
- But word order matters! "Dog bites man" ≠ "Man bites dog"

### Solution: Positional Encodings
Inject position information into the input embeddings.

### Types

**1. Sinusoidal (Original Transformer)**
$$PE_{(pos, 2i)} = \sin(pos / 10000^{2i/d})$$
$$PE_{(pos, 2i+1)} = \cos(pos / 10000^{2i/d})$$
- Fixed, no learned parameters
- Can theoretically generalize to unseen lengths

**2. Learned Positional Embeddings (BERT, GPT-2)**
- Lookup table: position → learned embedding
- Simple, effective, but fixed max length

**3. RoPE - Rotary Position Embedding (LLaMA, Mistral)**
- Rotate Q and K vectors based on position
- $q_m \cdot k_n = f(q, m) \cdot f(k, n)$ depends only on relative position $m-n$
- Captures relative position naturally in attention scores
- Can be extended via interpolation (PI, NTK, YaRN)

**4. ALiBi (Attention with Linear Biases)**
- No positional embedding at all
- Add linear bias to attention scores: $bias = -m \cdot |i - j|$
- Generalizes well to longer sequences

### Comparison
| Type | Relative Position | Length Generalization | Used In |
|------|-------------------|---------------------|--------|
| Sinusoidal | Implicit | Limited | Original Transformer |
| Learned | No | No | BERT, GPT-2 |
| RoPE | Yes | Yes (with scaling) | LLaMA, Mistral, Gemma |
| ALiBi | Yes | Yes | BLOOM, MPT |

## Key Takeaways
- Self-attention is **order-agnostic** — positional encodings provide sequence ordering
- **RoPE** is the modern standard — captures relative position, extensible context
- **ALiBi** is the simplest for length generalization (plain linear bias)
- Without positional info, transformers treat input as a **bag of tokens**
