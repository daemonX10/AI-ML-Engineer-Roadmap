# Q05: Inference Optimization
**Topic:** LLM | **Difficulty:** Hard | **Time:** 25 min
**Sheet:** GrindLLM50

---

## Question
How do you optimize inference?

## Detailed Answer

### 1. Model Optimization
- **Quantization**: FP16 → INT8 → INT4 (GPTQ, AWQ, GGUF)
- **Pruning**: Remove redundant weights/heads
- **Distillation**: Smaller student model
- **Speculative decoding**: Small draft model generates candidates, large model verifies in parallel

### 2. Attention Optimization
- **FlashAttention**: Fused, memory-efficient attention kernel
- **GQA/MQA**: Reduce KV cache size
- **PagedAttention (vLLM)**: Non-contiguous KV cache allocation
- **Continuous batching**: Add new requests to batch without waiting for others to finish

### 3. Batching Strategies
- **Static batching**: Wait for batch to fill → process together (high latency)
- **Continuous/Dynamic batching**: Process requests as they arrive, interleave generation
- **Iteration-level scheduling**: Each decoding step can add/remove sequences

### 4. Hardware Optimization
- **Tensor parallelism**: Split model across GPUs (reduce latency)
- **Pipeline parallelism**: Split layers across GPUs (increase throughput)
- **CUDA graphs**: Pre-compile GPU kernel sequences
- **Custom kernels**: Triton, TensorRT-LLM

### 5. Serving Frameworks
| Framework | Key Feature |
|-----------|------------|
| vLLM | PagedAttention, continuous batching |
| TensorRT-LLM | NVIDIA optimized kernels |
| TGI (HuggingFace) | Easy deployment |
| Ollama | Local/edge deployment |
| llama.cpp | CPU inference, GGUF format |

### 6. Speculative Decoding
1. Small model generates K draft tokens quickly
2. Large model verifies all K tokens in one forward pass
3. Accept correct tokens, resample from first incorrect position
4. Speedup: 2-3x with no quality loss

## Key Takeaways
- **Quantization** (INT4/INT8) gives the biggest single improvement
- **vLLM** with PagedAttention + continuous batching is the standard serving setup
- **Speculative decoding** gives 2-3x speedup with zero quality loss
- The stack: quantization + FlashAttention + GQA + continuous batching + tensor parallelism
