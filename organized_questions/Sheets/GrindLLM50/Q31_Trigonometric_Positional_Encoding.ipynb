## Question
Why do we use Trigonometric functions for Positional Encoding?

## Detailed Answer

### The Sinusoidal Encoding
$$PE_{(pos, 2i)} = \sin\left(\frac{pos}{10000^{2i/d}}\right)$$
$$PE_{(pos, 2i+1)} = \cos\left(\frac{pos}{10000^{2i/d}}\right)$$

### Why Trigonometric Functions?

**1. Relative Position via Linear Transformation**
For any fixed offset $k$, $PE_{pos+k}$ can be expressed as a **linear function** of $PE_{pos}$:
$$\begin{bmatrix} \sin(pos + k) \\ \cos(pos + k) \end{bmatrix} = \begin{bmatrix} \cos k & \sin k \\ -\sin k & \cos k \end{bmatrix} \begin{bmatrix} \sin(pos) \\ \cos(pos) \end{bmatrix}$$
This means the model can learn to attend to relative positions using simple linear operations.

**2. Bounded Values**
- $\sin, \cos \in [-1, 1]$ — values don't explode with position
- Uniform scale regardless of sequence length

**3. Unique Encoding per Position**
- Each position gets a unique pattern across d dimensions
- Different frequencies: low dimensions = high frequency (local), high dimensions = low frequency (global)
- Like a binary representation but smooth and continuous

**4. Generalization**
- Continuous functions → can extrapolate to positions not seen during training
- No learned parameters → works out-of-the-box for any length

**5. Orthogonal Properties**
- The dot product of PE vectors encodes relative distance
- $PE_{pos_1} \cdot PE_{pos_2}$ is a function of $|pos_1 - pos_2|$
- This helps attention learn relative-position patterns

### Multi-Scale Wavelengths
- Dimension 0: wavelength $2\pi$ (changes every position)
- Dimension d-1: wavelength $10000 \cdot 2\pi$ (changes very slowly)
- Like a clock: seconds hand (fast) vs hour hand (slow)

## Key Takeaways
- Trig functions enable **relative position via linear transformation**
- Bounded [-1, 1], unique per position, smooth for generalization
- Multi-scale wavelengths capture both local and global position info
- Modern alternative: **RoPE** applies rotation directly to Q, K (same principle)
