## Question
Explain how contextual embeddings are created in a Transformer.

## Detailed Answer

### Static vs Contextual Embeddings
| Type | Example | Same word = |
|------|---------|=============|
| Static | Word2Vec, GloVe | Same vector always |
| Contextual | BERT, GPT | Different vector per context |

Example: "bank" in "river bank" vs "bank account" → different embeddings

### How Contextual Embeddings Are Created

**Step 1: Token + Position Embedding**
$$h^0_i = \text{TokenEmbed}(x_i) + \text{PosEmbed}(i)$$
- Initial representation: context-independent (like static embeddings)

**Step 2: Self-Attention Layers (repeated L times)**
For each layer $l$:

a) **Multi-Head Self-Attention**:
$$\text{Attn}(h^{l-1}) = \text{softmax}\left(\frac{Q K^T}{\sqrt{d_k}}\right) V$$
- Each token "looks at" all other tokens (or past tokens for causal)
- Attention aggregates information from other positions
- This is where context gets mixed in!

b) **Feed-Forward Network**:
$$\text{FFN}(x) = \text{GELU}(xW_1 + b_1)W_2 + b_2$$
- Position-wise transformation (same for each token independently)
- Adds non-linear processing capacity

c) **Residual + LayerNorm**:
$$h^l_i = \text{LayerNorm}(h^{l-1}_i + \text{Attn}(h^{l-1}_i))$$
$$h^l_i = \text{LayerNorm}(h^l_i + \text{FFN}(h^l_i))$$

**Step 3: Final Layer Output**
After L layers, $h^L_i$ is a rich contextual embedding of token $i$.

### Why It Works
- Layer 1: Local syntactic patterns (nearby words)
- Middle layers: Semantic relationships, entities
- Final layers: Task-specific, high-level understanding
- Each attention layer progressively mixes in more context from other tokens
- After 12+ layers, each token's representation is influenced by the ENTIRE sequence

## Key Takeaways
- Contextual embeddings = token representations that **change based on surrounding context**
- Created through **stacked self-attention layers**: each layer mixes in more context
- Attention is the key mechanism — each token aggregates information from others
- Lower layers ≈ syntax, higher layers ≈ semantics/task-specific features
