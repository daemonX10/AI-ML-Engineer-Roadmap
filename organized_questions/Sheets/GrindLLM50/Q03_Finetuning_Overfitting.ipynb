# Q03: Fine-tuning Overfitting and Catastrophic Forgetting

## Question
Why might fine-tuning lead to overfitting or catastrophic forgetting?

## Detailed Answer

### Overfitting During Fine-tuning
**Why**: Fine-tuning datasets are typically small (100s-10Ks of examples) while the model has billions of parameters.

**Mitigation:**
- **Lower learning rate** (1e-5 to 5e-5 vs 1e-4 for pre-training)
- **LoRA/QLoRA**: Only train a small number of adapter parameters
- **Data augmentation**: Paraphrasing, back-translation
- **Early stopping**: Monitor validation loss
- **Dropout**: Keep 0.1 dropout during fine-tuning
- **Weight decay**: Regularization on weights

### Catastrophic Forgetting
**What**: After fine-tuning on task B, the model loses performance on the original task A.

**Why**: Gradient updates for new task overwrite weights important for old tasks.

**Mitigation:**
- **LoRA/PEFT**: Freeze base weights, only train adapters → base knowledge preserved
- **Elastic Weight Consolidation (EWC)**: Add regularization to penalize changing important weights
- **Mix in pre-training data**: Include some base task data during fine-tuning
- **Progressive training**: Gradually shift from old to new data
- **Multi-task learning**: Train on old + new tasks jointly
- **Lower learning rate**: Smaller updates → less disruption

### Best Practice
| Approach | Prevents Overfitting | Prevents Forgetting | Complexity |
|----------|---------------------|--------------------|-----------|
| LoRA | Yes (few params) | Yes (base frozen) | Low |
| Full fine-tune + low LR | Somewhat | Somewhat | Low |
| EWC | No | Yes | Medium |
| Mix pre-training data | No | Yes | Low |
| QLoRA | Yes | Yes | Low |

## Key Takeaways
- **LoRA/QLoRA** is the modern solution — prevents BOTH overfitting and forgetting
- Lower learning rate and early stopping are essential for full fine-tuning
- Mixing in pre-training data during fine-tuning helps maintain base capabilities
- The smaller the fine-tuning dataset, the higher the risk of both problems
