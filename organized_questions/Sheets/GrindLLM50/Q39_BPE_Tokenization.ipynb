# Q39: Byte-Pair Encoding (BPE) Tokenization
**Topic:** LLM | **Difficulty:** Medium | **Time:** 10 min
**Sheet:** GrindLLM50

---

## Question
How does Byte-Pair Encoding (BPE) tokenization work, with an example?

## Detailed Answer

### Algorithm
BPE builds a vocabulary by iteratively **merging the most frequent pair** of tokens.

### Step-by-Step Example
**Corpus**: `low low low low lowest lowest wider wider wider`

**Step 0**: Start with character-level tokens
```
Vocabulary: {l, o, w, e, s, t, i, d, r}
Tokenized: l-o-w (×4), l-o-w-e-s-t (×2), w-i-d-e-r (×3)
```

**Step 1**: Most frequent pair: (l, o) → 6 times → merge to `lo`
```
Vocabulary: {l, o, w, e, s, t, i, d, r, lo}
Tokenized: lo-w (×4), lo-w-e-s-t (×2), w-i-d-e-r (×3)
```

**Step 2**: Most frequent pair: (lo, w) → 6 times → merge to `low`
```
Vocabulary: {..., low}
Tokenized: low (×4), low-e-s-t (×2), w-i-d-e-r (×3)
```

**Step 3**: Most frequent pair: (e, r) → 3 times → merge to `er`
```
Vocabulary: {..., er}
Tokenized: low (×4), low-e-s-t (×2), w-i-d-er (×3)
```

Continue until desired vocabulary size is reached.

### Key Properties
- Common words → single token ("the", "and")
- Rare words → split into subwords ("anthropomorphize" → "anthrop", "omorph", "ize")
- **No out-of-vocabulary words** — everything can be tokenized
- Vocabulary size typically 32K-128K tokens

### Variants
| Method | Used In |
|--------|--------|
| BPE | GPT-2/3/4 |
| Byte-level BPE | GPT-2+ (operates on UTF-8 bytes) |
| SentencePiece (Unigram) | LLaMA, T5 |
| WordPiece | BERT |
| Tiktoken | OpenAI models |

## Key Takeaways
- BPE = iteratively merge most frequent character pairs until target vocab size
- Common words → single tokens, rare words → subword pieces
- **No OOV** (out-of-vocabulary) — any text can be tokenized
- Modern LLMs use byte-level BPE (GPT) or SentencePiece (LLaMA)
