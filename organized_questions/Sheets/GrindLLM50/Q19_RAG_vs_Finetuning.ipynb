# Q19: RAG vs Fine-tuning vs Prompting
**Topic:** LLM | **Difficulty:** Hard | **Time:** 25 min
**Sheet:** GrindLLM50

---

## Question
When to choose RAG vs Fine-tuning vs Prompting / PEFT (LoRA, QLoRA, adapters)?

## Detailed Answer

### Decision Framework

| Criteria | Prompting | RAG | Fine-tuning/PEFT |
|----------|-----------|-----|------------------|
| **Setup cost** | None | Medium | High |
| **Data needed** | 0 | Docs/KB | 100s-10Ks examples |
| **Latency** | Low | Medium | Low (after training) |
| **Up-to-date info** | No | Yes (update KB) | No (retrain needed) |
| **Domain adaptation** | Weak | Medium | Strong |
| **Hallucination control** | Moderate | Good (grounded) | Moderate |
| **Compute cost** | Low | Medium | High (training) |

### When to Use Each

**Prompting (Zero/Few-shot, Chain-of-Thought)**
- Simple tasks, well-defined format
- Quick prototyping / MVP
- No custom data available
- Tasks the base model already handles well

**RAG (Retrieval-Augmented Generation)**
- Need up-to-date or proprietary knowledge
- Large document corpus to query
- Need citations/source attribution
- Knowledge changes frequently
- Want to reduce hallucinations by grounding in documents

**Fine-tuning / PEFT**
- Need specific behavior/style/format the base model can't do via prompting
- Domain-specific vocabulary or reasoning (medical, legal, finance)
- Want to reduce inference cost (shorter prompts after fine-tuning)
- Classification tasks with specific label schema
- Quality requirements exceed what prompting can achieve

### Practical Decision Tree
```
1. Can prompting solve it? → YES → Use prompting
2. Do you need external/dynamic knowledge? → YES → Use RAG
3. Do you need behavior/style change? → YES → Fine-tune (LoRA/QLoRA)
4. Need both knowledge + behavior? → RAG + Fine-tuned model
```

### PEFT Methods Compared
| Method | Trainable Params | VRAM | Quality |
|--------|-----------------|------|--------|
| Full fine-tune | 100% | Very high | Best |
| LoRA | 0.1-1% | Medium | Very good |
| QLoRA | 0.1-1% | Low | Good |
| Prefix tuning | <0.1% | Low | OK |
| Prompt tuning | <0.01% | Minimal | Limited |

## Key Takeaways
- **Start with prompting** — cheapest, fastest to validate
- **RAG** when you need external/changing knowledge with source attribution
- **Fine-tuning** when you need behavior change or domain specialization
- **LoRA/QLoRA** makes fine-tuning accessible — 0.1% params, consumer GPU
- Most production systems combine approaches: fine-tuned model + RAG
