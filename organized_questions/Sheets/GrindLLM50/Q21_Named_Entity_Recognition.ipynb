## Question
What does Named Entity Recognition (NER) mean?

## Detailed Answer

### Definition
NER is the task of **identifying and classifying named entities** (specific real-world objects) in text into predefined categories.

### Common Entity Types
| Tag | Entity Type | Example |
|-----|------------|--------|
| PER | Person | "Albert Einstein" |
| ORG | Organization | "Google" |
| LOC | Location | "Paris" |
| DATE | Date/Time | "January 2024" |
| MONEY | Monetary | "$500 million" |

### BIO Tagging Scheme
```
"John Smith works at Google in New York"
 B-PER I-PER  O    O  B-ORG O B-LOC I-LOC
```
- **B-X**: Beginning of entity type X
- **I-X**: Inside (continuation of) entity type X
- **O**: Outside (not an entity)

### Approaches

**1. Traditional**: CRF (Conditional Random Fields), HMM + hand-crafted features

**2. Deep Learning**: BiLSTM-CRF (classic), BERT + token classification head

**3. Modern**: Fine-tuned LLM with token classification head:
- BERT-base + Linear → per-token prediction
- Output: softmax over BIO tags for each token
- Loss: Cross-entropy per token

**4. LLM-as-NER**: Prompt-based: "Extract all person names from: ..."
- Works well for simple cases, no fine-tuning needed
- Lower precision/recall than fine-tuned models

### Applications
- Search engines, knowledge graph construction, information extraction
- Medical NER (drug names, diseases), legal NER (parties, dates)

## Key Takeaways
- NER = **token-level classification** into entity categories
- **BIO tagging** distinguishes entity beginnings from continuations
- Modern approach: **BERT + linear head** fine-tuned on NER dataset
- Evaluated with **span-level F1** (exact match of entity boundaries + type)
