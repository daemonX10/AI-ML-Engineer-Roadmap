# Q24: Additive vs Multiplicative Attention
**Topic:** LLM | **Difficulty:** Medium | **Time:** 15 min
**Sheet:** GrindLLM50

---

## Question
Differences between multiplicative and additive attention. Pros and Cons? Do both give the same performance?

## Detailed Answer

### Additive Attention (Bahdanau)
$$e_{ij} = v_a^T \tanh(W_a [s_{i-1}; h_j])$$
- Concatenate query and key, pass through a small MLP
- Parameters: $W_a \in R^{d_a \times (d_q + d_k)}$, $v_a \in R^{d_a}$

### Multiplicative Attention (Luong / Scaled Dot-Product)
$$e_{ij} = \frac{q_i^T k_j}{\sqrt{d_k}}$$
- Simple dot product (optionally with scaling)
- No additional learned parameters (in basic form)

### Comparison
| Aspect | Additive | Multiplicative |
|--------|----------|----------------|
| **Computation** | MLP forward pass | Matrix multiplication |
| **Parameters** | Extra W_a, v_a | None (or W for general form) |
| **Speed** | Slower | Faster (parallelizable) |
| **GPU Efficiency** | Less efficient | Highly optimized (BLAS) |
| **Small d_k** | Better | Similar |
| **Large d_k** | Similar | Needs scaling (√d_k) |

### Performance
- For **small d_k**: Additive outperforms unscaled multiplicative attention
- For **large d_k** with scaling: Roughly equivalent performance
- The scaling factor $1/\sqrt{d_k}$ was introduced specifically to prevent softmax saturation with large d_k
- In practice, **scaled dot-product wins** due to computational efficiency

### Why Multiplicative Won
All modern Transformers use multiplicative (dot-product) attention because:
1. Matrix multiplications are **extremely optimized** on GPUs
2. No extra parameters to train
3. With scaling, performs equivalently to additive
4. FlashAttention is specifically designed for dot-product attention

## Key Takeaways
- Additive = MLP-based, multiplicative = dot-product-based
- **Performance is similar** when dot-product is properly scaled by $1/\sqrt{d_k}$
- Multiplicative dominates in practice due to **GPU parallelism** and hardware optimization
- All modern Transformers use scaled dot-product attention
