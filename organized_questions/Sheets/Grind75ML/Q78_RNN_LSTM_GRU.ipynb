## Question
Explain RNN, LSTM, and GRU. What problems do LSTM/GRU solve?

## Detailed Answer

### RNN (Recurrent Neural Network)
$$h_t = \tanh(W_{hh} h_{t-1} + W_{xh} x_t + b)$$
- Processes sequences step by step, maintaining hidden state
- **Problem**: Vanishing/exploding gradients for long sequences
- Can't capture long-range dependencies (gradient vanishes after ~10-20 steps)

### LSTM (Long Short-Term Memory)
Adds **gates** to control information flow:
- **Forget gate**: $f_t = \sigma(W_f [h_{t-1}, x_t] + b_f)$ — what to forget
- **Input gate**: $i_t = \sigma(W_i [h_{t-1}, x_t] + b_i)$ — what to store
- **Output gate**: $o_t = \sigma(W_o [h_{t-1}, x_t] + b_o)$ — what to output
- **Cell state**: $c_t = f_t \odot c_{t-1} + i_t \odot \tilde{c}_t$ — memory highway

The cell state provides a **highway** for gradients — they can flow through many timesteps without vanishing.

### GRU (Gated Recurrent Unit)
Simplified LSTM with 2 gates instead of 3:
- **Reset gate**: $r_t = \sigma(W_r [h_{t-1}, x_t])$
- **Update gate**: $z_t = \sigma(W_z [h_{t-1}, x_t])$
- $h_t = (1-z_t) \odot h_{t-1} + z_t \odot \tilde{h}_t$
- Fewer parameters than LSTM, often similar performance

### Comparison
| Feature | RNN | LSTM | GRU |
|---------|-----|------|-----|
| Gates | 0 | 3 | 2 |
| Parameters | Fewest | Most | Middle |
| Long-range | Poor | Good | Good |
| Speed | Fast | Slowest | Medium |
| Modern use | Rare | Some | Some |

**Note**: Transformers have largely replaced RNNs/LSTMs for most tasks.
