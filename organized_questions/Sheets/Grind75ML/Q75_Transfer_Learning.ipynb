## Question
What is Transfer Learning? When and how should you use it?

## Detailed Answer

### Definition
Use a model pre-trained on one task as initialization for a different (related) task.

### Why It Works
- Lower layers learn **general features** (edges, textures, n-grams)
- These features are **transferable** across tasks
- Higher layers are task-specific → need fine-tuning

### Strategies
| Strategy | When | How |
|----------|------|-----|
| **Feature extraction** | Very little data, similar task | Freeze all layers, train new head |
| **Fine-tune top layers** | Some data, related task | Freeze early layers, train later layers + head |
| **Fine-tune all** | Lots of data, different task | Initialize from pre-trained, train everything |
| **Progressive unfreezing** | Moderate data | Unfreeze layers gradually (ULMFiT approach) |

### Common Pre-trained Models
| Domain | Model | Pre-training Data |
|--------|-------|------------------|
| Vision | ResNet, EfficientNet | ImageNet (1.2M images) |
| Vision | CLIP, DINOv2 | Web-scale images |
| NLP | BERT | BooksCorpus + Wikipedia |
| NLP | GPT-2/3/4 | Web text |
| Multi-modal | CLIP | Image-text pairs |

### Best Practices
- **Lower learning rate** for pre-trained layers (10x-100x smaller)
- **Larger learning rate** for new head
- **Data augmentation** is essential for small datasets
- **Match preprocessing** to the pre-trained model (same normalization, resolution)

## Key Takeaways
- Understand the core concepts and trade-offs
- Be ready to discuss practical applications and alternatives
