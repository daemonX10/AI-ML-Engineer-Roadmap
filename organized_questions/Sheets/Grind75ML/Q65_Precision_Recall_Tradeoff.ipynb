## Question
What is the Precision-Recall tradeoff and how do you decide the threshold?

## Detailed Answer

### The Tradeoff
For any binary classifier outputting probabilities:
- **Increase threshold** → higher precision, lower recall (fewer predictions, but more accurate)
- **Decrease threshold** → lower precision, higher recall (more predictions, catching more positives)

### Why It Exists
$$\text{Precision} = \frac{TP}{TP+FP}, \quad \text{Recall} = \frac{TP}{TP+FN}$$
- Raising threshold reduces both TP and FP, but FP drops faster → precision ↑, recall ↓
- Lowering threshold increases both TP and FP, but FP grows faster → recall ↑, precision ↓

### How to Decide the Threshold

| Approach | Method |
|----------|--------|
| **Max F1** | Find threshold that maximizes F1 = $\frac{2PR}{P+R}$ |
| **Business rules** | Set minimum recall (e.g., detect 95% of fraud) |
| **Cost-based** | Minimize: $C_{FP} \cdot FP + C_{FN} \cdot FN$ |
| **PR curve** | Plot and pick operating point |

### Application-Specific Priorities
| Application | Prioritize | Why |
|------------|-----------|-----|
| Spam filter | Precision | Don't lose legitimate emails |
| Cancer screening | Recall | Don't miss cancer cases |
| Fraud detection | Recall | Catch most fraud |
| Search engine | Precision@K | Show relevant results first |
| Autonomous driving | Recall | Detect all obstacles |

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import precision_recall_curve, f1_score

np.random.seed(42)
y_true = np.array([1]*30 + [0]*70)
y_scores = np.concatenate([np.random.beta(5, 2, 30), np.random.beta(2, 5, 70)])

precision, recall, thresholds = precision_recall_curve(y_true, y_scores)
f1_scores = 2 * precision[:-1] * recall[:-1] / (precision[:-1] + recall[:-1] + 1e-8)
best_idx = np.argmax(f1_scores)

fig, ax = plt.subplots(figsize=(8, 5))
ax.plot(thresholds, precision[:-1], 'b-', label='Precision')
ax.plot(thresholds, recall[:-1], 'r-', label='Recall')
ax.plot(thresholds, f1_scores, 'g--', label='F1')
ax.axvline(thresholds[best_idx], color='k', linestyle=':', label=f'Best threshold={thresholds[best_idx]:.2f}')
ax.set_xlabel('Threshold')
ax.set_ylabel('Score')
ax.set_title('Precision-Recall Tradeoff')
ax.legend()
ax.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()
print(f'Best F1={f1_scores[best_idx]:.3f} at threshold={thresholds[best_idx]:.3f}')
```

## Key Takeaways
- **Always tune the threshold** — 0.5 is almost never optimal
- Use **F1-maximizing threshold** as default; adjust based on business needs
- **Cancer/fraud/safety**: optimize for recall; **spam/search**: optimize for precision
- Plot the **PR curve** — it tells the full story for imbalanced problems
