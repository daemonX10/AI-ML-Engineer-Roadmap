## Question
What is the difference between L1 and L2 regularization?

## Detailed Answer

### L1 Regularization (Lasso)
$$\mathcal{L} = \text{Loss} + \lambda \sum_i |w_i|$$
- Penalty proportional to **absolute value** of weights
- Produces **sparse** solutions (many weights become exactly 0)
- Acts as **feature selection** — irrelevant features get zero weight
- Gradient is constant (±λ) regardless of weight magnitude

### L2 Regularization (Ridge)
$$\mathcal{L} = \text{Loss} + \lambda \sum_i w_i^2$$
- Penalty proportional to **squared** weights
- Produces **small but non-zero** weights
- Distributes weight across correlated features
- Gradient is proportional to $w$ — large weights are penalized more

### Elastic Net (L1 + L2)
$$\mathcal{L} = \text{Loss} + \lambda_1 \sum |w_i| + \lambda_2 \sum w_i^2$$
Best of both: sparsity + stability with correlated features.

### Comparison
| Aspect | L1 (Lasso) | L2 (Ridge) |
|--------|-----------|------------|
| Sparsity | Yes (feature selection) | No |
| Correlated features | Picks one arbitrarily | Distributes weights |
| Solution uniqueness | May not be unique | Always unique |
| Computational | Harder (non-differentiable at 0) | Easier (smooth) |
| Geometry | Diamond constraint | Circle constraint |
| Deep learning | Rarely used | Weight decay (common) |

```python
import numpy as np
from sklearn.linear_model import Lasso, Ridge, ElasticNet
from sklearn.datasets import make_regression

X, y = make_regression(n_samples=100, n_features=20, n_informative=5, noise=10, random_state=42)

lasso = Lasso(alpha=0.5).fit(X, y)
ridge = Ridge(alpha=0.5).fit(X, y)
enet = ElasticNet(alpha=0.5, l1_ratio=0.5).fit(X, y)

print(f'Lasso: {np.sum(np.abs(lasso.coef_) < 1e-5)} zero coefficients out of {len(lasso.coef_)}')
print(f'Ridge: {np.sum(np.abs(ridge.coef_) < 1e-5)} zero coefficients out of {len(ridge.coef_)}')
print(f'ElasticNet: {np.sum(np.abs(enet.coef_) < 1e-5)} zero coefficients out of {len(enet.coef_)}')

print(f'\nLasso coefs: {np.round(lasso.coef_, 2)}')
print(f'Ridge coefs: {np.round(ridge.coef_, 2)}')
```

## Key Takeaways
- **L1** for feature selection (drives weights to zero)
- **L2** when you want to keep all features with small weights
- **ElasticNet** when you have correlated features and want some sparsity
- In deep learning: **weight decay** (L2 variant) is the standard; L1 is rarely used
