## Question
What distinguishes lemmatization from stemming?

## Detailed Answer

Both reduce words to their base form, but through different approaches:

### Stemming
**Rule-based** removal of suffixes to get the word stem.
- Fast, no linguistic knowledge needed
- Can produce non-real words ("studies" → "studi")
- Algorithms: Porter Stemmer, Snowball Stemmer, Lancaster Stemmer

Examples:
| Word | Stem |
|------|------|
| running | run |
| studies | studi |
| better | better |
| cats | cat |

### Lemmatization
**Dictionary-based** reduction to the word's lemma (dictionary form).
- Needs POS (part-of-speech) tag for accuracy
- Always produces real words
- Slower (dictionary lookup)
- Tools: WordNet Lemmatizer, spaCy

Examples:
| Word | POS | Lemma |
|------|-----|-------|
| running | verb | run |
| studies | verb | study |
| better | adj | good |
| cats | noun | cat |

### Comparison:
| Property | Stemming | Lemmatization |
|----------|----------|---------------|
| Speed | Fast | Slower |
| Output quality | May not be real words | Always real words |
| Context-aware | No | Yes (with POS) |
| Use case | Search engines, IR | Text analysis, NLP |
| Approach | Rule-based | Dictionary-based |

```python
# Stemming examples
from nltk.stem import PorterStemmer, SnowballStemmer
from nltk.stem import WordNetLemmatizer
import nltk
nltk.download('wordnet', quiet=True)

porter = PorterStemmer()
snowball = SnowballStemmer('english')
lemmatizer = WordNetLemmatizer()

words = ['running', 'studies', 'better', 'cats', 'geese', 'wolves', 'caring', 'fairly']

print(f'{"Word":<12} {"Porter":<12} {"Snowball":<12} {"Lemma(n)":<12} {"Lemma(v)":<12}')
print('-' * 60)
for w in words:
    print(f'{w:<12} {porter.stem(w):<12} {snowball.stem(w):<12} '
          f'{lemmatizer.lemmatize(w, "n"):<12} {lemmatizer.lemmatize(w, "v"):<12}')
```

## Key Takeaways
- **Stemming** is faster but cruder — use for search indexing and information retrieval
- **Lemmatization** produces real words — use when text quality matters (sentiment analysis, chatbots)
- Modern NLP (BERT, GPT) uses **subword tokenization** (BPE/WordPiece) instead — no need for either
- When using lemmatization, always provide the **POS tag** for correct results
