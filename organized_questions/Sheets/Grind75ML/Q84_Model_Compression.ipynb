## Question
What are the main techniques for model compression?

## Detailed Answer

### 1. Quantization
Reduce numerical precision: FP32 → FP16 → INT8 → INT4
- **Post-training quantization**: Quantize after training (easy, slight accuracy loss)
- **Quantization-aware training (QAT)**: Simulate quantization during training (better accuracy)
- **GPTQ, AWQ**: Advanced methods for LLM quantization
- 4-bit quantization: 4x memory reduction with minimal accuracy loss

### 2. Pruning
Remove redundant weights/neurons/attention heads.
- **Unstructured**: Remove individual weights (sparse matrices)
- **Structured**: Remove entire channels/layers (easier to accelerate)
- Typically 50-90% of weights can be pruned
- Lottery Ticket Hypothesis: Dense networks contain sparse subnetworks that train just as well

### 3. Knowledge Distillation
Train a small **student** model to mimic a large **teacher** model.
$$L = \alpha L_{CE}(y, p_s) + (1-\alpha) T^2 D_{KL}(p_t^{(T)}, p_s^{(T)})$$
- Student learns from teacher's soft probabilities (dark knowledge)
- DistilBERT: 60% of BERT's size, 97% of performance

### 4. Architecture Search / Efficient Architectures
- MobileNet: Depthwise separable convolutions
- EfficientNet: Compound scaling
- TinyBERT, MiniLM: Small-by-design language models

### Comparison
| Technique | Size Reduction | Accuracy Impact | Effort |
|-----------|---------------|-----------------|--------|
| INT8 Quant | 4x | Minimal | Low |
| INT4 Quant | 8x | Small | Medium |
| Pruning (50%) | 2x | Small | Medium |
| Distillation | Custom | Small-Medium | High |
