# Q50: Batch Normalization Problems
**Topic:** Deep-learning | **Difficulty:** Medium | **Time:** 8 min
**Sheet:** Grind75ML

---

## Question
What problems can you get with Batch Normalization?

## Detailed Answer

### Problem 1: Small Batch Size Sensitivity
BatchNorm statistics become noisy with small batches.
- Batch mean/variance are poor estimates when batch << 32
- Object detection/segmentation often use batch size 1-2 per GPU
- **Fix**: Use GroupNorm or LayerNorm instead

### Problem 2: Train-Test Discrepancy
- Training: Uses batch statistics (mean, var per batch)
- Inference: Uses running (exponential moving average) statistics
- If data distribution shifts, running stats may be stale
- **Fix**: Re-compute running stats, or use eval mode carefully

### Problem 3: Distributed Training Issues
Each GPU computes batch stats on its local mini-batch.
- With 8 GPUs and batch 256, each GPU sees only 32 samples
- Statistics are inconsistent across GPUs
- **Fix**: SyncBatchNorm (synchronize stats across GPUs)

### Problem 4: Breaks with Variable-Length Sequences
- Padding in RNNs/Transformers introduces zeros
- BatchNorm over padded positions corrupts statistics
- **Fix**: Use LayerNorm for sequence models

### Problem 5: Dependency Between Samples
- Each sample's normalized value depends on other samples in the batch
- This breaks assumptions in some architectures (e.g., GANs)
- Makes model behavior non-deterministic for single sample inference

### Problem 6: Information Leakage
- In Siamese/contrastive models, batch stats can leak information between positive/negative pairs
- **Fix**: Separate BN for each branch, or use LayerNorm

## Key Takeaways
- BatchNorm fails with **small batches** — use GroupNorm or LayerNorm
- **Train/test discrepancy** is a common bug — check running stats
- **Transformers use LayerNorm**, not BatchNorm, for good reasons
- Always use **SyncBatchNorm** in distributed training
