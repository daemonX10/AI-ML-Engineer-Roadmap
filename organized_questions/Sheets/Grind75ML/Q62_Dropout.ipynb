## Question
What is Dropout and how does it prevent overfitting?

## Detailed Answer

### What is Dropout?
Randomly set neuron activations to **zero** during training with probability $p$.

### How It Works
- **Training**: Each neuron is dropped with probability $p$ (typically 0.1-0.5)
- **Inference**: No dropout, but scale activations by $(1-p)$ (inverted dropout handles this during training)

### Why It Prevents Overfitting
1. **Ensemble effect**: Each training step uses a different "thinned" network → effectively trains $2^n$ different networks
2. **Co-adaptation prevention**: Neurons can't rely on specific other neurons → learn more robust features
3. **Noise injection**: Acts as data augmentation in feature space
4. **Implicit regularization**: Similar to L2 regularization (shown theoretically)

### Inverted Dropout (Standard Practice)
Scale activations by $\frac{1}{1-p}$ during training → no change needed at inference:
```python
mask = (np.random.rand(*h.shape) > p).astype(float)
h = h * mask / (1 - p)  # Scale up to maintain expected values
```

### Where to Apply
- **Large fully-connected layers**: Yes (p=0.5)
- **Conv layers**: Less common, use smaller p (0.1-0.2)
- **Transformers**: Applied to attention weights and FFN (p=0.1 typical)
- **Small networks**: Usually not needed (already limited capacity)

### Variants
- **Spatial Dropout**: Drop entire feature maps (for CNNs)
- **DropConnect**: Drop individual weights instead of neurons
- **DropPath/Stochastic Depth**: Drop entire layers (used in ResNets)

## Key Takeaways
- Dropout = **training-time ensemble** via random neuron zeroing
- Use **inverted dropout** (scale during training) — standard in all frameworks
- **p=0.1** for Transformers, **p=0.5** for large FC layers
- Remember to call `model.eval()` at inference — this disables dropout
