## Question
How to use one-class SVM for anomaly detection?

## Detailed Answer

### One-Class SVM
Train only on **normal** data. Learn a boundary that encloses most normal data → anything outside is an anomaly.

### How It Works
1. Map data to feature space using a kernel (typically RBF)
2. Find a hyperplane that separates data from the origin with maximum margin
3. Optimization:
$$\min_{w,\rho,\xi} \frac{1}{2}\|w\|^2 + \frac{1}{\nu n}\sum_i \xi_i - \rho$$
$$\text{s.t. } w \cdot \phi(x_i) \geq \rho - \xi_i, \quad \xi_i \geq 0$$

**Key parameter $\nu$**: Upper bound on the fraction of outliers, lower bound on fraction of support vectors.

### Parameters
| Parameter | Effect |
|-----------|--------|
| $\nu$ | Controls anomaly fraction (0.01 = expect 1% anomalies) |
| kernel | RBF is most common |
| $\gamma$ | RBF width (high = tight boundary, low = loose) |

### Advantages
- Only needs normal data for training (unsupervised)
- Non-parametric boundary shape (kernel trick)
- Works well in moderate dimensions

### Limitations
- Doesn't scale well to very large datasets (kernel matrix is $O(n^2)$)
- Sensitive to kernel/gamma choice
- Better alternatives for high-d: Isolation Forest, Autoencoder-based methods
