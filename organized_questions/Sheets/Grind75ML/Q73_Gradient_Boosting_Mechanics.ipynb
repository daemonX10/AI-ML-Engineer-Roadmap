## Question
Explain how Gradient Boosting works step by step.

## Detailed Answer

### Algorithm
1. Initialize with a constant: $F_0(x) = \arg\min_c \sum L(y_i, c)$
2. For $m = 1, ..., M$:
   a. Compute pseudo-residuals: $r_{im} = -\frac{\partial L(y_i, F_{m-1}(x_i))}{\partial F_{m-1}(x_i)}$
   b. Fit a tree $h_m$ to the pseudo-residuals
   c. Find optimal step size: $\gamma_m = \arg\min_\gamma \sum L(y_i, F_{m-1}(x_i) + \gamma h_m(x_i))$
   d. Update: $F_m(x) = F_{m-1}(x) + \nu \gamma_m h_m(x)$

Where $\nu$ is the learning rate (shrinkage).

### For MSE Loss
$$L = \frac{1}{2}(y - \hat{y})^2 \implies r_i = y_i - \hat{y}_i$$
Pseudo-residuals = actual residuals! Each tree corrects previous tree's errors.

### For Log Loss (Classification)
$$L = -[y\log(p) + (1-y)\log(1-p)] \implies r_i = y_i - p_i$$
Still fits residuals, but prediction is passed through sigmoid.

### Why "Gradient" Boosting?
Each step = gradient descent in **function space**.
- Instead of updating parameters, we add a new function (tree)
- The tree approximates the negative gradient of the loss
- This works for ANY differentiable loss function

### Regularization Knobs
| Parameter | Effect |
|-----------|--------|
| Learning rate ($\nu$) | Smaller → more trees needed, better generalization |
| n_estimators | More trees → more fitting power (risk of overfit) |
| max_depth | Shallow trees (3-5) = weak learners → better |
| subsample | Row subsampling → stochastic GB (like RF's bagging) |
| colsample | Feature subsampling → more diverse trees |
