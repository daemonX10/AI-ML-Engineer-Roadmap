## Question
What are Exploding Gradients?

## Detailed Answer

### Definition
Exploding gradients occur when gradients grow exponentially large during backpropagation, causing:
- Extremely large weight updates
- NaN/Inf values in loss and parameters
- Training instability or divergence

### Why It Happens
In deep networks, gradients are computed via chain rule:
$$\frac{\partial L}{\partial W_1} = \frac{\partial L}{\partial h_n} \prod_{i=1}^{n-1} \frac{\partial h_{i+1}}{\partial h_i}$$

If $\|\frac{\partial h_{i+1}}{\partial h_i}\| > 1$ for many layers, the product grows **exponentially**.

### Most Common In:
- Deep RNNs (long sequences → many multiplications)
- Very deep networks without residual connections
- Poor weight initialization (too large)
- High learning rates

### Detection
- Loss suddenly becomes NaN or Inf
- Weights become very large
- Training loss oscillates wildly

### Solutions
| Solution | How It Helps |
|----------|-------------|
| **Gradient clipping** | Cap gradient norm: `torch.nn.utils.clip_grad_norm_(model.parameters(), max_norm)` |
| **Proper initialization** | He/Xavier keep gradient magnitudes stable |
| **Batch/Layer normalization** | Stabilize activations → stable gradients |
| **Skip connections (ResNets)** | Provide gradient highway, reducing long chain products |
| **LSTM/GRU gates** | Gating mechanisms control gradient flow in RNNs |
| **Lower learning rate** | Smaller updates even with large gradients |
| **Gradient scaling (mixed precision)** | Scale loss up to prevent underflow, then unscale gradients |

### Gradient Clipping (Most Common Fix)
```python
# Clip by global norm (most common)
torch.nn.utils.clip_grad_norm_(model.parameters(), max_norm=1.0)

# Clip by value
torch.nn.utils.clip_grad_value_(model.parameters(), clip_value=0.5)
```
Standard in Transformer training — nearly every LLM uses gradient clipping.

## Key Takeaways
- Exploding gradients = exponential growth in backpropagated gradients
- **Gradient clipping** is the standard fix — used in virtually all Transformer training
- **Skip connections** (ResNet, Transformers) prevent gradient vanishing/exploding
- LSTM gates were specifically designed to control gradient flow in RNNs
