## Question
What is the significance of breadth vs depth in a neural network?

## Detailed Answer

### Depth (Number of Layers)
- Enables **hierarchical feature learning**: simple → complex representations
- Each layer learns a higher level of abstraction
- Deeper networks can represent more complex functions with fewer total parameters
- **Universal Approximation**: A 2-layer network can approximate any function, but may need exponentially many neurons

### Breadth (Neurons per Layer)
- More neurons = more features captured at each level
- Wide networks are better at **memorization**
- Easier to train (less vanishing gradient)
- Over-parameterized wide networks generalize surprisingly well (double descent)

### Key Results

**Depth is more efficient:**
- A depth-$k$ network can represent functions requiring exponentially many neurons in a depth-$(k-1)$ network
- ResNets (152 layers) dramatically outperformed wider shallow networks

**Width matters too:**
- Neural Tangent Kernel (NTK) theory: infinite-width networks behave like kernel methods
- Wide ResNets showed width helps for smaller models
- GPT-3 scaled both depth (96 layers) and width (12288 hidden)

### Practical Guidelines
| Scenario | Recommendation |
|----------|---------------|
| Limited compute | Deeper + narrower (parameter efficient) |
| Parallelization | Wider (easier to parallelize across neurons) |
| Tabular data | Few wide layers (2-3 layers, 256-512 neurons) |
| Images | Deep (ResNet, EfficientNet) |
| Transformers | Scale both together (scaling laws) |

## Key Takeaways
- **Depth enables hierarchical representations** — more expressive per parameter
- **Width helps with optimization** — easier training, better parallelization
- In practice: **scale both** (scaling laws show power-law improvements)
- ResNets showed depth is feasible with skip connections; Transformers scale both
