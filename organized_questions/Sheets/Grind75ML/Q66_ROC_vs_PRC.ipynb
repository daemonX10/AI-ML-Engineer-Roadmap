# Q66: ROC Curve vs Precision-Recall Curve

## Question
When should you use ROC curve vs Precision-Recall curve?

## Detailed Answer

### ROC Curve
Plots TPR vs FPR. AUC=0.5 is random, 1.0 is perfect. Good for balanced datasets but misleading for imbalanced data.

### PR Curve
Plots Precision vs Recall. Better for imbalanced datasets since it focuses on positive class performance.

### When to Use Which
| Scenario | Use |
|----------|-----|
| Balanced | ROC-AUC |
| Imbalanced | PR-AUC |
| Care about negatives | ROC |
| Care about positives | PR |

## Key Takeaways
- Use **PR-AUC for imbalanced data** -- ROC can be misleadingly high
- ROC-AUC is fine for balanced problems
- Always provide both curves for complete picture
