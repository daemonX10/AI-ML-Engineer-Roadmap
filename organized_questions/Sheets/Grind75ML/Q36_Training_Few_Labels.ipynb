# Q36: Training with Few Labels

**Sheet:** Grind75ML

## Question
Suppose you want to train a classifier with plenty of unlabeled training data but only a few thousand labeled instances, how would you proceed?

## Detailed Answer

### Strategy 1: Semi-Supervised Learning
Use both labeled and unlabeled data together.
- **Pseudo-labeling**: Train on labeled data, predict unlabeled, add confident predictions as labels
- **FixMatch**: Weak augmentation for pseudo-labels + strong augmentation for training
- **MixMatch**: Combines consistency regularization, pseudo-labels, and MixUp

### Strategy 2: Self-Supervised Pre-training
Pre-train on unlabeled data with a pretext task, then fine-tune on labeled data.
- **SimCLR / MoCo**: Contrastive learning on image augmentations
- **DINO / MAE**: Vision transformers with self-supervised objectives
- **BERT**: Masked language modeling (NLP)

### Strategy 3: Transfer Learning
Use a model pre-trained on a large dataset (ImageNet, COCO) as initialization.
- Freeze early layers, fine-tune later layers
- Works even with <100 labeled examples

### Strategy 4: Data Augmentation
Synthetically increase labeled data.
- **Image**: Random crop, flip, color jitter, CutMix, MixUp, RandAugment
- **Text**: Back-translation, synonym replacement, paraphrasing
- **Both**: Generative augmentation with diffusion/LLM

### Strategy 5: Active Learning
Intelligently select which samples to label next.
- Query the most uncertain / informative samples
- Human labels only the most valuable examples
- Can achieve 90% of full-data performance with 10-20% of labels

### Strategy 6: Few-Shot Learning
- **Meta-learning** (MAML): Learn to learn from few examples
- **Prototypical networks**: Compare new examples to class prototypes

### Recommended Pipeline
```
1. Pre-train with self-supervised learning on all unlabeled data
2. Fine-tune on labeled data with strong augmentation
3. Generate pseudo-labels for unlabeled data
4. Re-train on labeled + pseudo-labeled data
5. Use active learning to select next batch to label
```

## Key Takeaways
- Understanding the underlying principles is more important than memorizing formulas
- Always consider practical implications alongside theoretical knowledge
- Be ready to discuss trade-offs and when to apply each approach
