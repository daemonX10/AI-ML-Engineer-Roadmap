## Question
Explain the Transformer architecture. Why has it replaced RNNs?

## Detailed Answer

### Architecture
```
Encoder:                          Decoder:
Input Embeddings + PE             Output Embeddings + PE
  ↓                                 ↓
Multi-Head Self-Attention         Masked Multi-Head Self-Attention
  ↓ (Add & Norm)                   ↓ (Add & Norm)
Feed-Forward Network              Multi-Head Cross-Attention
  ↓ (Add & Norm)                   ↓ (Add & Norm)
  × N layers                      Feed-Forward Network
                                    ↓ (Add & Norm)
                                    × N layers
```

### Key Components
1. **Multi-Head Self-Attention**: Each position attends to all others (see Q76)
2. **Positional Encoding**: Since there's no recurrence, position info must be added
   - Sinusoidal: $PE_{pos, 2i} = \sin(pos / 10000^{2i/d})$
   - RoPE (modern): Rotary positional embeddings in LLaMA, GPT-NeoX
3. **Feed-Forward Network**: Two linear layers with non-linearity (GELU)
   - $FFN(x) = \text{GELU}(xW_1 + b_1)W_2 + b_2$
4. **Layer Normalization**: Stabilizes training
5. **Residual Connections**: Around every sub-layer

### Why Transformers Replaced RNNs
| Advantage | Explanation |
|-----------|-------------|
| **Parallelization** | All positions processed simultaneously (RNN is sequential) |
| **Long-range dependencies** | Attention is O(1) path length (RNN is O(n)) |
| **Scalability** | Scales efficiently to billions of parameters |
| **Training speed** | Much faster on GPUs (parallel computation) |

### Variants
- **Encoder-only**: BERT (bidirectional, classification/understanding)
- **Decoder-only**: GPT (autoregressive, generation)
- **Encoder-Decoder**: T5, BART (seq2seq, translation)
