# Q30: How do you select the Activation Function?

## Question
How do you select the Activation Function for a Deep Learning Model?

## Detailed Answer

### Selection Guidelines

#### Output Layer (depends on task)
| Task | Activation | Reason |
|------|-----------|--------|
| Binary classification | Sigmoid | Output in (0,1)=probability |
| Multi-class classification | Softmax | Probabilities that sum to 1 |
| Regression | None (linear) | Unbounded output |
| Bounded regression | Sigmoid/Tanh | Output in (0,1) or (-1,1) |
| Multi-label | Sigmoid (per class) | Independent probabilities |

#### Hidden Layers
| Architecture | Recommended | Why |
|-------------|-------------|-----|
| CNN (ResNet, etc.) | ReLU | Fast, prevents vanishing gradient |
| Transformer (BERT, GPT) | GELU | Smooth, slightly better empirically |
| RNN/LSTM gates | Sigmoid + Tanh | Gates need (0,1), cell state needs (-1,1) |
| GAN Generator | LeakyReLU/ReLU | Avoid dead neurons in generator |
| GAN Discriminator | LeakyReLU | Better gradient flow |
| Efficiency-focused | ReLU | Cheapest to compute |
| Swish/SiLU | EfficientNet, Mamba | Better accuracy than ReLU |

### Decision Flowchart
```
Is it the output layer?
├── Yes → What's the task?
│   ├── Binary → Sigmoid
│   ├── Multi-class → Softmax
│   └── Regression → Linear
└── No → What's the architecture?
    ├── Transformer → GELU
    ├── CNN → ReLU (or Swish for SOTA)
    ├── RNN gate → Sigmoid/Tanh
    └── Default → ReLU
```

### Common Mistakes
- Using sigmoid in hidden layers → vanishing gradients
- Using ReLU in output for classification → not a probability
- Not pairing activation with proper weight initialization

## Key Takeaways
- **Output activation** is determined by the task; **hidden activation** by the architecture
- When in doubt: **ReLU for hidden layers** — it's the safest default
- Pair initialization with activation: ReLU → He init, Sigmoid/Tanh → Xavier init
- Modern trend: GELU is replacing ReLU in state-of-the-art architectures
