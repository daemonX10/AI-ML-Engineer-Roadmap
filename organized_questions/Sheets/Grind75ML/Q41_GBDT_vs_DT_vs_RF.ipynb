## Question
What distinguishes a gradient boost decision tree (GBDT), a decision tree, and a random forest?

## Detailed Answer

### Decision Tree (DT)
Single tree that recursively splits data based on feature thresholds.
- Simple, interpretable
- Prone to overfitting (high variance)
- Unstable — small data changes → different tree

### Random Forest (RF)
**Ensemble of DTs trained independently** via bagging.
- Each tree trained on a random subset of data (bootstrap sampling)
- Each split considers random subset of features
- Final prediction = majority vote (classification) or average (regression)
- **Parallel** training — trees are independent

### Gradient Boosted Decision Trees (GBDT)
**Ensemble of DTs trained sequentially** — each tree corrects the errors of previous trees.
- Each tree fits the **residuals** (gradient of loss) of the current ensemble
- Final prediction = sum of all trees (weighted)
- **Sequential** training — each tree depends on previous

### Comparison
| Feature | DT | RF | GBDT |
|---------|----|----|------|
| Bias | Low | Low | Low |
| Variance | High | Low (bagging) | Low (boosting) |
| Interpretability | High | Medium | Low |
| Training speed | Fast | Parallelizable | Sequential (slower) |
| Overfitting risk | High | Low | Medium (needs tuning) |
| Typical performance | Baseline | Good | Best (tabular data SOTA) |
| Key library | sklearn | sklearn | XGBoost, LightGBM, CatBoost |

### When to Use What
- **DT**: Interpretability needed, small data, baselines
- **RF**: General-purpose, minimal tuning, robust
- **GBDT**: Maximum accuracy on tabular data, willing to tune hyperparameters

## Key Takeaways
- Understanding the underlying principles is more important than memorizing formulas
- Always consider practical implications alongside theoretical knowledge
- Be ready to discuss trade-offs and when to apply each approach
