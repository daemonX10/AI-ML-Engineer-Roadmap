# Q61: Batch vs Mini-batch vs Stochastic GD

## Question
What are the differences between Batch GD, Mini-batch GD, and Stochastic GD?

## Detailed Answer

### Batch Gradient Descent
Uses **entire dataset** for each gradient update.
$$\theta = \theta - \alpha \cdot \frac{1}{N} \sum_{i=1}^{N} \nabla L(x_i, y_i; \theta)$$
- Stable convergence, smooth loss curve
- Very slow for large datasets
- Requires entire dataset in memory

### Stochastic Gradient Descent (SGD)
Uses **single sample** for each update.
$$\theta = \theta - \alpha \cdot \nabla L(x_i, y_i; \theta)$$
- Very noisy gradients — helps escape local minima
- Fast per update, but many updates needed
- Can't leverage vectorized operations efficiently

### Mini-batch Gradient Descent
Uses **small batch** (e.g., 32-512 samples) for each update.
$$\theta = \theta - \alpha \cdot \frac{1}{B} \sum_{i=1}^{B} \nabla L(x_i, y_i; \theta)$$
- Best of both worlds: noise for exploration + vectorization for speed
- **This is what everyone uses in practice** (often just called "SGD")

### Comparison
| Feature | Batch | Mini-batch | SGD |
|---------|-------|-----------|-----|
| Samples per update | N (all) | B (32-512) | 1 |
| Speed per epoch | Slow | Fast | Medium |
| Gradient noise | None | Low-Med | High |
| Memory | O(N) | O(B) | O(1) |
| Convergence | Smooth | Balanced | Noisy |
| GPU utilization | Max | Good | Poor |

### Typical Batch Sizes
| Domain | Typical B | Reason |
|--------|----------|--------|
| Computer Vision | 32-256 | GPU memory limits |
| NLP / Transformers | 8-64 per GPU | Sequence length × batch = memory |
| LLM Training | Millions (effective) | Gradient accumulation |

## Key Takeaways
- **Mini-batch GD** is the standard — everyone uses it
- Batch size is a hyperparameter: larger = more stable, smaller = more regularization
- Gradient accumulation lets you simulate large batches with limited GPU memory
- 'SGD' in practice almost always means mini-batch SGD
