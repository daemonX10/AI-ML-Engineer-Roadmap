## Question
Why are Non-linear Activation functions important?

## Detailed Answer

### Without Non-linearity
If every layer uses a linear activation (or no activation), the entire network collapses to a single linear transformation:

$$f(x) = W_3(W_2(W_1 x)) = (W_3 W_2 W_1) x = W' x$$

No matter how many layers, you still have a **linear model** — it can only learn linear decision boundaries.

### With Non-linearity
Non-linear activation functions allow the network to learn **complex, non-linear mappings**:

$$f(x) = \sigma_3(W_3 \cdot \sigma_2(W_2 \cdot \sigma_1(W_1 x)))$$

This gives neural networks the power of **universal function approximation** (can approximate any continuous function given enough neurons).

### Common Activation Functions

| Function | Formula | Range | Pros | Cons |
|----------|---------|-------|------|------|
| **Sigmoid** | $\frac{1}{1+e^{-x}}$ | (0,1) | Smooth probability | Vanishing gradients, not zero-centered |
| **Tanh** | $\frac{e^x-e^{-x}}{e^x+e^{-x}}$ | (-1,1) | Zero-centered | Vanishing gradients |
| **ReLU** | $\max(0,x)$ | [0,∞) | Fast, no vanishing grad | Dead neurons (dying ReLU) |
| **Leaky ReLU** | $\max(0.01x, x)$ | (-∞,∞) | No dead neurons | Small negative slope is arbitrary |
| **ELU** | $x$ if $x>0$, $\alpha(e^x-1)$ | (-α,∞) | Smooth, pushes mean→0 | Slower (exp) |
| **GELU** | $x \cdot \Phi(x)$ | ≈(-0.17,∞) | Smooth, stochastic | Slower |
| **Swish/SiLU** | $x \cdot \sigma(x)$ | ≈(-0.28,∞) | Outperforms ReLU | Slightly slower |

### Key Properties for Activation Functions
1. **Non-linearity**: Essential for learning complex patterns
2. **Differentiable**: Required for gradient-based optimization (backpropagation)
3. **Non-saturating**: Avoids vanishing gradients (ReLU family)
4. **Zero-centered**: Helps optimization (tanh, but not sigmoid/ReLU)

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-5, 5, 200)

activations = {
    'Sigmoid': (1/(1+np.exp(-x)), 'blue'),
    'Tanh': (np.tanh(x), 'orange'),
    'ReLU': (np.maximum(0, x), 'green'),
    'Leaky ReLU': (np.where(x > 0, x, 0.01*x), 'red'),
    'GELU': (x * 0.5*(1 + np.vectorize(lambda z: np.tanh(np.sqrt(2/np.pi)*(z + 0.044715*z**3)))(x)), 'purple'),
    'Swish': (x / (1+np.exp(-x)), 'brown'),
}

fig, axes = plt.subplots(2, 3, figsize=(15, 8))
for ax, (name, (vals, color)) in zip(axes.flat, activations.items()):
    ax.plot(x, vals, color=color, linewidth=2)
    ax.axhline(0, color='k', linewidth=0.5)
    ax.axvline(0, color='k', linewidth=0.5)
    ax.set_title(name, fontsize=13)
    ax.grid(True, alpha=0.3)
    ax.set_xlim(-5, 5)
plt.tight_layout()
plt.show()
```

## Key Takeaways
- Without non-linearity, deep networks are just **linear models** — depth is useless
- **ReLU** is the default for hidden layers — simple, fast, no vanishing gradient
- **GELU** is the default for Transformers (BERT, GPT)
- Use **sigmoid/softmax** only at the output layer for probability outputs
