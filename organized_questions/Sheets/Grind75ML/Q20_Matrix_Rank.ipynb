## Question
What is the resultant rank of multiplying a 3×1 and 1×3 matrix?

## Detailed Answer

### Setup
Let $A$ be a $3 \times 1$ matrix (column vector) and $B$ be a $1 \times 3$ matrix (row vector):

$$A = \begin{bmatrix} a_1 \\ a_2 \\ a_3 \end{bmatrix}, \quad B = \begin{bmatrix} b_1 & b_2 & b_3 \end{bmatrix}$$

### Product
$$C = AB = \begin{bmatrix} a_1 b_1 & a_1 b_2 & a_1 b_3 \\ a_2 b_1 & a_2 b_2 & a_2 b_3 \\ a_3 b_1 & a_3 b_2 & a_3 b_3 \end{bmatrix}$$

This is a $3 \times 3$ matrix where every row is a scalar multiple of $B$.

### Rank Analysis
- Row 1 = $a_1 \cdot B$
- Row 2 = $a_2 \cdot B$  
- Row 3 = $a_3 \cdot B$

All rows are linearly dependent (scalar multiples of the same row vector $B$).

**Rank of $C$ = 1** (assuming $A \neq 0$ and $B \neq 0$)

### General Rule
$$\text{rank}(AB) \leq \min(\text{rank}(A), \text{rank}(B))$$

Since $\text{rank}(A) = 1$ (3×1 column) and $\text{rank}(B) = 1$ (1×3 row):
$$\text{rank}(AB) \leq \min(1, 1) = 1$$

And $\text{rank}(AB) = 1$ when neither $A$ nor $B$ is the zero vector.

### Outer Product
This product $AB$ is called an **outer product** — it always produces a rank-1 matrix. In fact, **any rank-1 matrix can be written as an outer product** $uv^T$.

```python
import numpy as np

# Demonstrate
A = np.array([[2], [3], [5]])   # 3x1
B = np.array([[1, 4, 7]])        # 1x3

C = A @ B  # Outer product: 3x3

print(f'A (3x1):\n{A}\n')
print(f'B (1x3):\n{B}\n')
print(f'C = A @ B (3x3):\n{C}\n')
print(f'Rank of C: {np.linalg.matrix_rank(C)}')
print(f'\nNote: Row 2 = {A[1,0]/A[0,0]:.1f} * Row 1')
print(f'      Row 3 = {A[2,0]/A[0,0]:.1f} * Row 1')

# Edge case: zero vector
A_zero = np.array([[0], [0], [0]])
C_zero = A_zero @ B
print(f'\nRank when A=0: {np.linalg.matrix_rank(C_zero)}')
```

## Key Takeaways
- Outer product of any two non-zero vectors always gives a **rank-1 matrix**
- $\text{rank}(AB) \leq \min(\text{rank}(A), \text{rank}(B))$ — fundamental rank inequality
- Rank-1 matrices are used in **LoRA** (Low-Rank Adaptation) for efficient LLM fine-tuning
- SVD decomposes any matrix into a sum of rank-1 matrices: $A = \sum \sigma_i u_i v_i^T$
