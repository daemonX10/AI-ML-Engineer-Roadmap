## Question
How does XGBoost work internally? What makes it faster than vanilla GBDT?

## Detailed Answer

### Core Idea
XGBoost = eXtreme Gradient Boosting. Builds trees sequentially, each fitting the **gradient of the loss** (negative residuals).

### Objective Function
$$\mathcal{L} = \sum_{i} l(y_i, \hat{y}_i^{(t)}) + \sum_{k} \Omega(f_k)$$
$$\Omega(f) = \gamma T + \frac{1}{2}\lambda \sum_{j=1}^{T} w_j^2$$

Uses **2nd-order Taylor expansion** of the loss:
$$\mathcal{L}^{(t)} \approx \sum_i [g_i f_t(x_i) + \frac{1}{2}h_i f_t^2(x_i)] + \Omega(f_t)$$
where $g_i = \partial l / \partial \hat{y}^{(t-1)}$, $h_i = \partial^2 l / \partial (\hat{y}^{(t-1)})^2$

### What Makes XGBoost Fast

| Innovation | How It Helps |
|-----------|-------------|
| **Approximate split finding** | Weighted quantile sketch for candidate splits (no sorting) |
| **Histogram-based** | Bin features → faster split evaluation |
| **Column block** | Pre-sorted compressed column format for parallel access |
| **Cache-aware access** | Optimized memory layout for tree traversal |
| **Out-of-core** | Disk-based computation for huge datasets |
| **Sparsity-aware** | Native handling of missing values (learns best default direction) |
| **Regularization** | L1 + L2 on leaf weights + tree complexity ($\gamma T$) |
| **Shrinkage** | Learning rate reduces each tree's contribution |
| **Column subsampling** | Like RF — reduces correlation between trees |

### Key Hyperparameters
| Parameter | What | Typical |
|-----------|------|--------|
| n_estimators | Number of trees | 100-1000 |
| max_depth | Tree depth | 3-8 |
| learning_rate (eta) | Shrinkage | 0.01-0.3 |
| subsample | Row sampling rate | 0.7-1.0 |
| colsample_bytree | Feature sampling | 0.7-1.0 |
| reg_alpha (L1) | L1 on weights | 0-1 |
| reg_lambda (L2) | L2 on weights | 1, tune |
