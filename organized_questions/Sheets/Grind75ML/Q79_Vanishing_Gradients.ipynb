## Question
What are Vanishing Gradients and how to mitigate them?

## Detailed Answer

### The Problem
In deep networks, gradients shrink exponentially during backpropagation:
$$\frac{\partial L}{\partial W_1} = \frac{\partial L}{\partial h_n} \prod_{i=1}^{n-1} \frac{\partial h_{i+1}}{\partial h_i}$$

If each gradient factor < 1, the product → 0 for deep layers.

### Why It Happens
- **Sigmoid/tanh saturation**: Gradients near 0 for large/small inputs
- Sigmoid: max gradient = 0.25 (at x=0); multiply many < 1 values → vanishes
- **Poor initialization**: Weights too small or too large

### Solutions

| Solution | How It Helps |
|----------|-------------|
| **ReLU activation** | Gradient = 1 for positive inputs (doesn't shrink) |
| **Residual connections** | $h_{l+1} = h_l + F(h_l)$ → gradient = 1 + ... |
| **Batch/Layer normalization** | Keeps activations in the non-saturating range |
| **LSTM/GRU** | Gating mechanism provides gradient highway |
| **Proper initialization** | Xavier/He keep gradient magnitudes stable |
| **Gradient clipping** | More for exploding, but prevents instability |

### The ResNet Innovation
$$y = F(x) + x$$
$$\frac{\partial y}{\partial x} = \frac{\partial F}{\partial x} + 1$$

The "+1" ensures gradients never vanish completely — they're always at least 1.
This is why ResNets can be 100+ layers deep.
