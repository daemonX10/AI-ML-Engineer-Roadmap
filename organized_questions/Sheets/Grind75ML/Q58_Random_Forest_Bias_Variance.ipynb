## Question
What is the relationship between random forests and the bias-variance tradeoff?

## Detailed Answer

### Decision Trees: Low Bias, High Variance
- A single deep tree overfits easily (high variance)
- Very small changes in data → completely different tree

### Random Forest: Reduces Variance via Bagging

**Bagging principle**: Average of M models reduces variance:
$$\text{Var}(\bar{f}) = \frac{\rho \sigma^2 + (1-\rho)\sigma^2 / M}{1}$$
where $\rho$ = correlation between trees, $\sigma^2$ = variance of single tree.

**Random feature subsets** reduce $\rho$:
- Each tree only considers $\sqrt{d}$ features at each split
- Trees become less correlated → more variance reduction

### The Result
- **Bias**: Same as individual tree (low) — RF doesn't reduce bias
- **Variance**: Greatly reduced via averaging decorrelated trees
- Net effect: Much better generalization

### Key Parameters
| Parameter | Effect on Bias | Effect on Variance |
|-----------|---------------|-------------------|
| n_estimators (trees) | No effect | Decreases (more = better) |
| max_depth | Increases (if shallow) | Decreases (if shallow) |
| max_features | Increases (fewer) | Decreases (fewer, more diverse) |
| min_samples_leaf | Increases | Decreases |

### RF vs Boosting (Bias Reduction)
- **RF** (bagging): Reduces **variance** — averages many low-bias, high-variance trees
- **GBDT** (boosting): Reduces **bias** — sequentially corrects errors
- Boosting can overfit more easily; RF is more robust out-of-the-box

## Key Takeaways
- Random Forest = averaging **decorrelated** trees → **variance reduction**
- Feature subsampling ($\sqrt{d}$) is key to decorrelation
- More trees always helps (or plateaus) — never hurts
- If you need bias reduction → use GBDT (XGBoost/LightGBM)
