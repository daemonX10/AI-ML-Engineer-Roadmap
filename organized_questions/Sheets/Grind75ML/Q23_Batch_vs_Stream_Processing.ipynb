# Q23: Batch Processing vs Stream Processing
**Topic:** MLOps | **Difficulty:** Medium | **Time:** 5 min
**Sheet:** Grind75ML

---

## Question
What's the difference between Batch Processing and Stream Processing?

## Detailed Answer

### Batch Processing
Process large volumes of data at scheduled intervals (hourly, daily).

**Characteristics:**
- High throughput, high latency
- Process all data in one go
- Fault tolerance through re-runs
- Tools: **Apache Spark**, Hadoop MapReduce, dbt, Airflow

**ML Use Cases:**
- Nightly model retraining
- Feature computation for feature store
- Batch inference (scoring entire databases)
- A/B test analysis

### Stream Processing
Process data continuously as it arrives (milliseconds to seconds latency).

**Characteristics:**
- Low throughput per event, low latency
- Process one event / micro-batch at a time
- Complex fault tolerance (exactly-once semantics)
- Tools: **Apache Kafka + Flink**, Spark Streaming, AWS Kinesis

**ML Use Cases:**
- Real-time fraud detection
- Live recommendation updates
- Real-time feature computation (click-through rate)
- Anomaly detection on sensor data

### Comparison
| Aspect | Batch | Stream |
|--------|-------|--------|
| Latency | Minutes–hours | Milliseconds–seconds |
| Throughput | High | Lower per event |
| Data size | Bounded (finite) | Unbounded (infinite) |
| Complexity | Simple | Complex (state management, ordering) |
| Cost | Lower (scheduled) | Higher (always-on) |
| Fault handling | Re-run job | Checkpointing, exactly-once delivery |
| Tools | Spark, Airflow, dbt | Kafka, Flink, Kinesis |

### Lambda vs Kappa Architecture
- **Lambda**: Dual path — batch layer (historical) + speed layer (real-time) + serving layer
- **Kappa**: Single path — everything through streaming; batch = replay of stream
- Modern trend: Kappa with Kafka + Flink (simpler to maintain)

## Key Takeaways
- Use **batch** for training, feature store backfills, and non-time-sensitive inference
- Use **stream** for real-time inference triggers, live features, and anomaly detection
- Most production ML systems use **both** — batch for training, stream for serving
- **Feature stores** (Feast, Tecton) bridge batch and stream by materializing features
