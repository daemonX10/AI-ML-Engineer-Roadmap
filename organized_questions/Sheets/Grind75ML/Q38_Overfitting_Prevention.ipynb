## Question
Why do models overfit and how do we prevent it?

## Detailed Answer

### Why Models Overfit
Overfitting occurs when a model learns **noise and patterns specific to training data** rather than generalizable patterns.

**Causes:**
- Model too complex relative to data (high capacity)
- Too few training samples
- Training too long (too many epochs)
- Noisy or mislabeled data

**Signs:** Training loss ↓ but validation loss ↑, train accuracy >> validation accuracy.

### Prevention Strategies

| Strategy | How | When |
|----------|-----|------|
| **More data** | Collect or augment | Always the best fix |
| **Data augmentation** | Flips, crops, MixUp | Images, limited data |
| **Regularization (L1/L2)** | Penalty on weight magnitude | Linear, tree models |
| **Dropout** | Randomly zero neurons during training | Neural networks |
| **Early stopping** | Stop when val loss stops improving | All iterative models |
| **Reduce model size** | Fewer layers/neurons | When model is too large |
| **Batch normalization** | Normalize layer inputs | DNNs |
| **Cross-validation** | K-fold for model selection | All models |
| **Ensemble methods** | Bagging, averaging | Production models |
| **Transfer learning** | Pre-trained weights | Limited labeled data |
| **Weight decay** | Decay weights each step | Transformer training |

### Bias-Variance Perspective
- **Overfitting** = low bias, high variance → model is too sensitive to training noise
- **Underfitting** = high bias, low variance → model is too simple
- Goal: find the sweet spot (optimal model complexity)

## Key Takeaways
- Understanding the underlying principles is more important than memorizing formulas
- Always consider practical implications alongside theoretical knowledge
- Be ready to discuss trade-offs and when to apply each approach
