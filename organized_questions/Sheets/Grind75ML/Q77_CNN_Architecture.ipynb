## Question
Explain the architecture of a CNN: convolutions, pooling, and fully connected layers.

## Detailed Answer

### Convolutional Layer
- Applies **learnable filters** (kernels) that slide across the input
- Each filter produces a **feature map** via dot product: $\sum_{i,j} I(x+i, y+j) \cdot K(i,j) + b$
- **Parameters**: kernel size, stride, padding, number of filters
- **Weight sharing**: Same filter applied everywhere → fewer parameters, translation equivariance
- Output size: $\frac{W - K + 2P}{S} + 1$

### Pooling Layer
- Reduces spatial dimensions (downsampling)
- **Max pooling**: Takes maximum value in each window (most common)
- **Average pooling**: Takes average
- **Global average pooling**: Average over entire spatial dimension → 1D vector
- No learnable parameters
- Provides slight translation invariance

### Fully Connected (FC) Layer
- Flatten spatial features → dense vector
- Standard neural network layers for classification
- Modern architectures often replace FC with **Global Average Pooling** (fewer parameters)

### Typical Architecture Flow
```
Input → [Conv → ReLU → Pool] × N → Flatten → FC → Softmax
```

### Key Architectures Evolution
| Architecture | Year | Key Innovation |
|-------------|------|----------------|
| LeNet-5 | 1998 | First practical CNN |
| AlexNet | 2012 | ReLU, dropout, GPU training |
| VGG | 2014 | Small filters (3×3), deep |
| Inception | 2014 | Multi-scale filters |
| ResNet | 2015 | Skip connections, 152 layers |
| EfficientNet | 2019 | Compound scaling |
| ConvNeXt | 2022 | CNN with Transformer tricks |
