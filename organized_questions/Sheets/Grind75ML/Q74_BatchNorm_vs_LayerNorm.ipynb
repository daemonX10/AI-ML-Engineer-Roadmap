# Q74: BatchNorm vs LayerNorm
**Topic:** Deep-learning | **Difficulty:** Medium | **Time:** 8 min
**Sheet:** Grind75ML

## Question
Compare Batch Normalization vs Layer Normalization. When to use which?

## Detailed Answer

### Batch Normalization
- Normalizes across the **batch** dimension for each feature
- For input shape (B, C, H, W): stats computed over (B, H, W) for each C
- **Batch-dependent**: Different behavior at train vs inference
- Works well with **large batch sizes** (>16)

### Layer Normalization
- Normalizes across the **feature** dimension for each sample
- For input shape (B, C, H, W): stats computed over (C, H, W) for each B
- **Batch-independent**: Same behavior at train and inference
- Works with **any batch size** (even 1)

### Comparison
| Feature | BatchNorm | LayerNorm |
|---------|-----------|----------|
| Normalize over | Batch | Features |
| Batch size dependent | Yes | No |
| Train/test difference | Yes (running stats) | No |
| Best for | CNNs | Transformers, RNNs |
| Sequence models | Bad (padding, variable length) | Good |
| Small batches | Bad | Good |
| Self-supervised | Can leak info | Safe |

### Why Transformers Use LayerNorm
1. Variable-length sequences → can't normalize across batch (padding)
2. Batch sizes are small (memory-limited by sequence length)
3. Pre-LayerNorm (normalize before attention) improves training stability
4. No train/test discrepancy
