## Question
What is the impact of Learning Rate on model training? What are LR schedules?

## Detailed Answer

### Learning Rate Impact
- **Too high**: Training diverges (loss increases or oscillates wildly)
- **Too low**: Training is extremely slow, may get stuck in suboptimal minima
- **Just right**: Fast convergence to good solution

### Common Learning Rate Schedules

| Schedule | Description | Used In |
|----------|-------------|--------|
| **Step decay** | Multiply LR by γ every N epochs | ResNets |
| **Cosine annealing** | $\eta_t = \eta_{min} + \frac{1}{2}(\eta_{max}-\eta_{min})(1+\cos(\frac{t}{T}\pi))$ | Transformers, modern CNNs |
| **Linear warmup** | Linearly increase LR from 0 to max over N steps | Transformers (first ~1K steps) |
| **Warmup + cosine** | Linear warmup → cosine decay | GPT, BERT |
| **Reduce on plateau** | Reduce LR when val loss stops improving | General training |
| **1cycle** | Warmup → high LR → cooldown in one cycle | Fast.ai (super convergence) |
| **Polynomial decay** | $\eta_t = \eta_0(1-t/T)^p$ | Various |

### Warmup
Start with very small LR, linearly increase to target LR over N steps.
- Prevents initial training instability (large random gradients at start)
- **Essential for Transformers** — Adam's moving averages need time to warm up

### Finding the Right LR
- **LR Range Test** (Leslie Smith): Gradually increase LR, plot loss. Use LR where loss decreases fastest
- **Typical values**: 1e-3 for Adam, 1e-1 for SGD, 5e-5 for fine-tuning BERT

## Key Takeaways
- LR is **the most important hyperparameter** — always tune it first
- **Warmup + cosine decay** is the modern default for Transformers
- Use LR range test to find the right starting value
- Adam needs lower LR than SGD; fine-tuning needs lower LR than from-scratch
