## Question
What techniques can be used for Feature Engineering with text data?

## Detailed Answer

### 1. Bag of Words (BoW)
Represent text as a vector of word counts. Vocabulary size = vector dimension.
- Simple but loses word order and semantics
- Sparse representation

### 2. TF-IDF (Term Frequency-Inverse Document Frequency)
$$\text{TF-IDF}(t,d) = \text{TF}(t,d) \times \log\frac{N}{\text{DF}(t)}$$
- Weighs rare words higher, common words lower
- Better than BoW for classification tasks

### 3. N-grams
Capture word order by considering sequences of N consecutive words:
- Unigrams: ("cat"), ("sat")
- Bigrams: ("the cat"), ("cat sat")
- Trigrams: ("the cat sat")

### 4. Word Embeddings (Dense)
- **Word2Vec** (Skip-gram / CBOW): 100-300d dense vectors, captures semantic similarity
- **GloVe**: Co-occurrence matrix factorization
- **FastText**: Subword embeddings, handles OOV words

### 5. Sentence/Document Embeddings
- **Sentence-BERT**: Fine-tuned BERT for sentence similarity
- **Doc2Vec**: Extension of Word2Vec to documents
- **Universal Sentence Encoder**: Google's general-purpose sentence embedding

### 6. Contextual Embeddings
- **BERT / RoBERTa / DeBERTa**: Context-dependent word representations
- Same word gets different embeddings in different contexts
- Current SOTA for most NLP tasks

### 7. Statistical Features
| Feature | Description |
|---------|-------------|
| Text length | Number of characters/words |
| Avg word length | Mean characters per word |
| Punctuation count | Special character usage |
| Capital ratio | Uppercase characters / total |
| Stop word ratio | Common words / total |
| Sentiment score | Polarity using lexicons |
| Readability | Flesch-Kincaid score |

### 8. Named Entities
Extract person, organization, location, date entities as features.

### When to Use What:
| Method | Best For | Dimensionality |
|--------|----------|----------------|
| BoW/TF-IDF | Small datasets, baselines | High (sparse) |
| Word2Vec/GloVe | Medium datasets, similarity | 100-300 (dense) |
| BERT embeddings | SOTA performance, ample compute | 768 (dense) |
| Statistical | Complement to any method | Low |

```python
import numpy as np
from sklearn.feature_extraction.text import CountVectorizer, TfidfVectorizer

# Sample corpus
corpus = [
    'The cat sat on the mat.',
    'The dog chased the cat.',
    'Dogs and cats are pets.',
    'The mat is on the floor.',
]

# --- Bag of Words ---
bow = CountVectorizer()
bow_matrix = bow.fit_transform(corpus)
print('=== Bag of Words ===')
print(f'Vocabulary: {bow.get_feature_names_out()}')
print(f'BOW Matrix:\n{bow_matrix.toarray()}\n')

# --- TF-IDF ---
tfidf = TfidfVectorizer()
tfidf_matrix = tfidf.fit_transform(corpus)
print('=== TF-IDF ===')
print(f'Vocabulary: {tfidf.get_feature_names_out()}')
print(f'TF-IDF Matrix (rounded):\n{np.round(tfidf_matrix.toarray(), 3)}\n')

# --- N-grams ---
bigram = CountVectorizer(ngram_range=(2, 2))
bigram_matrix = bigram.fit_transform(corpus)
print('=== Bigrams ===')
print(f'Bigram features: {bigram.get_feature_names_out()}')

# --- Statistical Features ---
def extract_text_features(text: str) -> dict:
    words = text.split()
    return {
        'char_count': len(text),
        'word_count': len(words),
        'avg_word_len': np.mean([len(w) for w in words]),
        'capital_ratio': sum(1 for c in text if c.isupper()) / len(text),
        'punct_count': sum(1 for c in text if not c.isalnum() and not c.isspace()),
    }

print('\n=== Statistical Features ===')
for doc in corpus[:2]:
    print(f'{doc} → {extract_text_features(doc)}')
```

## Key Takeaways
- Start with **TF-IDF** as a strong baseline — it's simple and works well
- **BERT embeddings** are SOTA but require GPU; use for production systems with compute budget
- **Combine** statistical features with embeddings for best results
- Always normalize text first: lowercasing, removing punctuation, stopword handling
