## Question
Explain how SVM works. What are support vectors?

## Detailed Answer

### The Goal
Find the hyperplane that **maximizes the margin** between classes.

### Formulation
$$\min_{w,b} \frac{1}{2}\|w\|^2 \quad \text{s.t. } y_i(w \cdot x_i + b) \geq 1$$

The margin width = $\frac{2}{\|w\|}$, so minimizing $\|w\|$ maximizes margin.

### Support Vectors
- Points that lie **exactly on the margin boundary** ($y_i(w \cdot x_i + b) = 1$)
- They are the **critical** data points — removing any non-SV doesn't change the decision boundary
- Typically a small subset of training data
- The decision boundary is completely determined by support vectors

### Dual Formulation
$$\max_\alpha \sum_i \alpha_i - \frac{1}{2}\sum_{i,j}\alpha_i \alpha_j y_i y_j (x_i \cdot x_j)$$
$$\text{s.t. } \alpha_i \geq 0, \sum_i \alpha_i y_i = 0$$

- Only $\alpha_i > 0$ for support vectors (rest are 0)
- The dual form only uses dot products → enables kernel trick

### Prediction
$$f(x) = \text{sign}\left(\sum_{\text{SVs}} \alpha_i y_i K(x_i, x) + b\right)$$

### Pros and Cons
| Pros | Cons |
|------|------|
| Effective in high dimensions | Slow for large N (O(N²) to O(N³)) |
| Memory efficient (only stores SVs) | Sensitive to feature scaling |
| Kernel trick for non-linear | Need to choose kernel and tune C, γ |
| Good generalization (max margin) | No direct probability output |
