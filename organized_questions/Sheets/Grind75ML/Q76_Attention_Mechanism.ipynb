## Question
Explain the Attention Mechanism. What is Self-Attention?

## Detailed Answer

### Attention — Core Idea
**Attend to relevant parts** of the input when producing output. Instead of compressing entire input into one vector, the model can look back at all input positions.

### Scaled Dot-Product Attention
$$\text{Attention}(Q, K, V) = \text{softmax}\left(\frac{QK^T}{\sqrt{d_k}}\right) V$$

- **Q** (Query): What I'm looking for
- **K** (Key): What I have to offer (index)
- **V** (Value): The actual content
- $\sqrt{d_k}$: Scaling factor to prevent softmax saturation

### Self-Attention
When Q, K, V all come from the **same sequence**:
$$Q = XW_Q, \quad K = XW_K, \quad V = XW_V$$
Each position attends to all other positions in the same sequence.

### Multi-Head Attention
Run $h$ attention heads in parallel with different projections:
$$\text{MultiHead}(Q,K,V) = \text{Concat}(\text{head}_1,...,\text{head}_h)W_O$$
$$\text{head}_i = \text{Attention}(QW_i^Q, KW_i^K, VW_i^V)$$
- Each head can attend to different relationship patterns
- One might capture syntactic relations, another semantic

### Types of Attention
| Type | Use |
|------|-----|
| Self-attention | Encoder (BERT), looks at full sequence |
| Causal/masked self-attention | Decoder (GPT), can only look left |
| Cross-attention | Decoder attending to encoder output |
