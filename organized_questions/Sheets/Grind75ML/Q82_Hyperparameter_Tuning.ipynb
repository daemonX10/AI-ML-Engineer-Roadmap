## Question
What are the best strategies for hyperparameter tuning?

## Detailed Answer

### Methods (in order of sophistication)

**1. Grid Search**
- Try all combinations of specified values
- Complete but exponential: $O(k^d)$ for k values per d parameters
- Good for ≤3 hyperparameters with known good ranges

**2. Random Search**
- Sample random combinations
- **Better than grid search** (Bergstra & Bengio 2012): covers the important dimensions more efficiently
- Easy to parallelize

**3. Bayesian Optimization**
- Build a probabilistic model (GP, TPE) of the objective function
- Use acquisition function (EI, UCB) to decide next point to evaluate
- Libraries: Optuna, HyperOpt, SMAC
- Most sample-efficient method

**4. Successive Halving / Hyperband**
- Start many configurations with small budget
- Keep best half, double the budget, repeat
- Very efficient when most configurations are clearly bad early

**5. Population-Based Training (PBT)**
- Evolve hyperparameters during training
- Used by DeepMind for RL agents

### Practical Recommendations
| Scenario | Method |
|----------|--------|
| Quick exploration | Random search |
| Expensive evaluations | Bayesian (Optuna) |
| Many hyperparameters | Hyperband + Bayesian |
| Neural network | Learning rate finder + Optuna |
