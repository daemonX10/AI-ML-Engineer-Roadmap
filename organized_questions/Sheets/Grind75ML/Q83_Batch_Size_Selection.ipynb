## Question
How do you select the batch size for training? What are the trade-offs?

## Detailed Answer

### Trade-offs
| Small Batch (8-64) | Large Batch (256-4096+) |
|--------------------|-----------------------|
| More noise → better generalization | Less noise → faster convergence |
| Less GPU utilization | Better GPU utilization |
| Lower memory | Higher memory |
| More updates per epoch | Fewer updates per epoch |
| Implicit regularization | May need explicit regularization |

### Key Findings
- **Small batches generalize better** (noise helps escape sharp minima)
- **Large batches converge to sharper minima** (poorer generalization)
- **Linear scaling rule**: If you multiply batch size by k, multiply LR by k
- **Warmup** is essential for large batch training

### Practical Guidelines
| Domain | Typical Batch Size |
|--------|-------------------|
| Image classification | 32-256 |
| Object detection | 2-16 per GPU |
| NLP/Transformers | 8-64 per GPU |
| LLM pre-training | Millions (with gradient accumulation) |
