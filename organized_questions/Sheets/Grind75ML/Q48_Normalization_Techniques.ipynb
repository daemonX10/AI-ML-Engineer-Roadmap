## Question
What are the different types of normalization techniques used in DNN?

## Detailed Answer

### 1. Batch Normalization (BatchNorm)
Normalize across the **batch dimension** for each feature:
$$\hat{x}_i = \frac{x_i - \mu_B}{\sqrt{\sigma_B^2 + \epsilon}} \quad \text{then} \quad y_i = \gamma \hat{x}_i + \beta$$
- $\mu_B$, $\sigma_B$: batch mean and variance
- $\gamma$, $\beta$: learnable scale and shift
- **Pros**: Faster training, regularization effect, higher learning rates
- **Cons**: Depends on batch size, different behavior at train/test, issues with RNNs

### 2. Layer Normalization (LayerNorm)
Normalize across the **feature dimension** for each sample:
- Independent of batch size
- **Default for Transformers** (BERT, GPT)
- Works well with RNNs (sequence models)

### 3. Instance Normalization
Normalize each **sample and channel** independently:
- Used in **style transfer** and image generation
- Removes instance-specific contrast information

### 4. Group Normalization
Divide channels into groups, normalize within each group:
- **Independent of batch size** (like LayerNorm)
- Better than BatchNorm for small batch sizes (object detection, segmentation)

### 5. RMSNorm
Simplified LayerNorm — only divides by RMS, no mean subtraction:
$$\hat{x}_i = \frac{x_i}{\text{RMS}(x)} \cdot \gamma \quad \text{where } \text{RMS}(x) = \sqrt{\frac{1}{d}\sum_i x_i^2}$$
- **Used in LLaMA, Gemma** — faster and equally effective

### Comparison
| Technique | Normalize Over | Best For | Batch-Dependent? |
|-----------|---------------|----------|------------------|
| BatchNorm | Batch | CNNs | Yes |
| LayerNorm | Features | Transformers, RNNs | No |
| InstanceNorm | H, W per channel | Style transfer | No |
| GroupNorm | Channel groups | Small batch CNN | No |
| RMSNorm | Features (no mean) | Modern LLMs | No |

## Key Takeaways
- **BatchNorm** for CNNs (large batch), **LayerNorm** for Transformers
- **GroupNorm** when batch size is small (detection/segmentation)
- **RMSNorm** is the modern trend for LLMs — simpler and faster than LayerNorm
- At inference, BatchNorm uses running statistics — LayerNorm doesn't change
