## Question
What is Cross-Validation and why is it important in supervised learning?

## Detailed Answer

**Cross-validation** evaluates model performance by splitting data into multiple train/test partitions.

### K-Fold Cross-Validation
1. Split data into K equal folds
2. For each fold i: train on K-1 folds, test on fold i
3. Average the K performance scores

### Variants
| Type | Description | Best For |
|------|-------------|----------|
| K-Fold | Standard K partitions | General use |
| Stratified K-Fold | Preserves class ratio in each fold | Imbalanced data |
| Leave-One-Out (LOO) | K = N, test on 1 sample | Very small datasets |
| Repeated K-Fold | Run K-fold multiple times | Stable estimates |
| Time Series Split | Expanding window, no future leakage | Temporal data |
| Group K-Fold | Keep groups together | Grouped data (patients, users) |

### Why Important
- **Reduces overfitting** to a single train/test split
- **Better estimate** of generalization performance
- Uses **all data** for both training and evaluation
- Essential for **hyperparameter tuning** (CV score = objective)

### Guidelines
- K=5 or K=10 is standard (5 is faster, 10 is more accurate)
- Always use **Stratified** K-Fold for classification
- For time series, **never** use random splits — use temporal splits

```python
import numpy as np
from sklearn.model_selection import cross_val_score, KFold, StratifiedKFold
from sklearn.datasets import make_classification
from sklearn.ensemble import RandomForestClassifier

X, y = make_classification(n_samples=1000, n_features=20, random_state=42)
rf = RandomForestClassifier(n_estimators=50, random_state=42)

# K-Fold
kf = KFold(n_splits=5, shuffle=True, random_state=42)
scores_kf = cross_val_score(rf, X, y, cv=kf, scoring='accuracy')
print(f'K-Fold (K=5): {scores_kf.mean():.4f} ± {scores_kf.std():.4f}')

# Stratified K-Fold
skf = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)
scores_skf = cross_val_score(rf, X, y, cv=skf, scoring='accuracy')
print(f'Stratified K-Fold: {scores_skf.mean():.4f} ± {scores_skf.std():.4f}')

print(f'\nPer-fold scores: {scores_skf.round(4)}')
```

## Key Takeaways
- Use **K=5** for fast iteration, **K=10** for final evaluation
- **Stratified** K-Fold is a must for classification — always preserves class distribution
- Never use random CV for time series — use `TimeSeriesSplit`
- CV error is the most reliable metric for model selection
