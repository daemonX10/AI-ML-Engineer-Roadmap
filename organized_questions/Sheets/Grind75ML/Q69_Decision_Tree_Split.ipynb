## Question
What splitting criteria does a Decision Tree use? Gini vs Entropy vs MSE.

## Detailed Answer

### Gini Impurity
gen_phase8b.pyG = 1 - \sum p_k^2gen_phase8b.py
Range [0, 0.5] for binary. Faster (no log). Default in sklearn.

### Entropy
gen_phase8b.pyH = -\sum p_k \log_2(p_k)gen_phase8b.py
Range [0, 1] for binary. Tends to produce balanced trees.

### Regression: MSE
gen_phase8b.pyMSE = \frac{1}{N}\sum(y_i - \bar{y})^2gen_phase8b.py
Minimize within-leaf variance. Leaf prediction = mean.

In practice, Gini and entropy produce nearly identical results.

## Key Takeaways
- Gini and entropy give nearly identical trees
- MSE is default for regression trees
- Be able to compute Gini/entropy manually in interviews
