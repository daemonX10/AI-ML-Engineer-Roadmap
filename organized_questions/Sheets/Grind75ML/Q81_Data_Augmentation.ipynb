## Question
What Data Augmentation techniques do you use for images and text?

## Detailed Answer

### Image Augmentation

**Basic Transforms:**
- Random horizontal/vertical flip
- Random rotation (±15°)
- Random crop and resize
- Color jitter (brightness, contrast, saturation, hue)
- Gaussian blur, noise

**Advanced Techniques:**
| Technique | Description | Used In |
|-----------|-------------|--------|
| **CutOut** | Random rectangular mask | DNN training |
| **MixUp** | Blend two images: $\tilde{x} = \lambda x_1 + (1-\lambda)x_2$ | All |
| **CutMix** | Paste patch from one image to another | Classification |
| **RandAugment** | Random sequence of N transforms with magnitude M | EfficientNet, ViT |
| **AutoAugment** | Learned augmentation policy | AutoML |
| **Mosaic** | Combine 4 images | YOLO |

### Text Augmentation
- **Synonym replacement**: Replace words with synonyms
- **Random insertion/deletion/swap**: Perturb word positions
- **Back-translation**: Translate to another language and back
- **Paraphrasing with LLM**: Use GPT to rephrase
- **Token masking**: BERT-style random token replacement

### Guidelines
- Augmentation that **preserves the label** — don't flip a "6" into a "9"
- For medical images: be careful with geometric transforms (pathology location matters)
- **Test-time augmentation (TTA)**: Augment at inference, average predictions
