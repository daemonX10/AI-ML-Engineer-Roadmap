## Question
How does Gaussian Mixture Model work? What is Expectation Maximization?

## Detailed Answer

### Gaussian Mixture Model (GMM)
Models data as a mixture of K Gaussian distributions:
$$p(x) = \sum_{k=1}^{K} \pi_k \mathcal{N}(x | \mu_k, \Sigma_k)$$
- $\pi_k$: mixing coefficient (weight of component k)
- $\mu_k$: mean of component k
- $\Sigma_k$: covariance matrix of component k

### Expectation Maximization (EM) Algorithm
Iterative algorithm to find maximum likelihood parameters when there are latent (hidden) variables.

**E-step**: Compute responsibility — probability that point $x_i$ belongs to cluster $k$:
$$r_{ik} = \frac{\pi_k \mathcal{N}(x_i | \mu_k, \Sigma_k)}{\sum_j \pi_j \mathcal{N}(x_i | \mu_j, \Sigma_j)}$$

**M-step**: Update parameters using responsibilities:
$$N_k = \sum_i r_{ik}, \quad \mu_k = \frac{1}{N_k}\sum_i r_{ik} x_i$$
$$\Sigma_k = \frac{1}{N_k}\sum_i r_{ik}(x_i - \mu_k)(x_i - \mu_k)^T, \quad \pi_k = \frac{N_k}{N}$$

**Repeat** E and M steps until convergence (log-likelihood stops increasing).

### GMM vs K-Means
| Feature | K-Means | GMM |
|---------|---------|-----|
| Assignment | Hard (0 or 1) | Soft (probability) |
| Cluster shape | Spherical only | Elliptical (any covariance) |
| Output | Cluster labels | Probabilities per cluster |
| K selection | Elbow/silhouette | BIC/AIC |
| Algorithm | Lloyd's | EM |

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.mixture import GaussianMixture
from sklearn.datasets import make_blobs

X, y_true = make_blobs(n_samples=300, centers=3, cluster_std=[1.0, 1.5, 0.5], random_state=42)

# Fit GMM
gmm = GaussianMixture(n_components=3, covariance_type='full', random_state=42)
y_gmm = gmm.fit_predict(X)
probs = gmm.predict_proba(X)

print(f'Means:\n{gmm.means_.round(2)}')
print(f'\nWeights: {gmm.weights_.round(3)}')
print(f'BIC: {gmm.bic(X):.1f}, AIC: {gmm.aic(X):.1f}')

# Soft assignment example
print(f'\nSample 0 probabilities: {probs[0].round(4)}')

plt.figure(figsize=(8, 6))
plt.scatter(X[:, 0], X[:, 1], c=y_gmm, cmap='viridis', s=20, alpha=0.7)
plt.scatter(gmm.means_[:, 0], gmm.means_[:, 1], c='red', marker='X', s=200, label='Centers')
plt.title('GMM Clustering')
plt.legend()
plt.show()
```

## Key Takeaways
- GMM = **soft clustering** with Gaussian assumptions → more flexible than K-Means
- EM alternates between computing responsibilities (E) and updating parameters (M)
- Use **BIC/AIC** for automatic K selection — major advantage over K-Means
- EM is a general algorithm — also used in HMMs, mixed-effects models, etc.
