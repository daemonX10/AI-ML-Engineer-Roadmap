## Question
While performing K-Means Clustering, how do you determine the value of K?

## Detailed Answer

### Method 1: Elbow Method
Plot inertia (within-cluster sum of squares) vs K. Look for the "elbow" where the rate of decrease slows down.
- Simple and intuitive
- Subjective — the elbow isn't always clear

### Method 2: Silhouette Score
For each point, measures how similar it is to its own cluster vs nearest other cluster:
$$s(i) = \frac{b(i) - a(i)}{\max(a(i), b(i))}$$
- Range: [-1, 1]. Higher = better clustering
- Choose K that maximizes average silhouette score

### Method 3: Gap Statistic
Compare within-cluster dispersion to that of a null reference distribution (uniform random):
$$\text{Gap}(k) = E[\log W_k^*] - \log W_k$$
- Choose smallest K where Gap(k) ≥ Gap(k+1) - s(k+1)

### Method 4: Information Criteria
- **BIC / AIC**: Based on likelihood + penalty for model complexity
- Works well with Gaussian Mixture Models (GMM) as an alternative to K-Means

### Method 5: Domain Knowledge
- Customer segments: business knows they want 3-5 groups
- Image compression: K = number of desired colors

### Recommendation
Use **Silhouette Score** as primary + **Elbow Method** as visual confirmation. For production, prefer **GMM with BIC** for automatic model selection.

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.datasets import make_blobs
from sklearn.metrics import silhouette_score

X, _ = make_blobs(n_samples=500, centers=4, cluster_std=1.0, random_state=42)

K_range = range(2, 10)
inertias = []
silhouettes = []

for k in K_range:
    km = KMeans(n_clusters=k, random_state=42, n_init=10)
    km.fit(X)
    inertias.append(km.inertia_)
    silhouettes.append(silhouette_score(X, km.labels_))

fig, axes = plt.subplots(1, 2, figsize=(14, 5))
axes[0].plot(K_range, inertias, 'bo-')
axes[0].set_xlabel('K'); axes[0].set_ylabel('Inertia')
axes[0].set_title('Elbow Method')
axes[0].grid(True, alpha=0.3)

axes[1].plot(K_range, silhouettes, 'ro-')
axes[1].set_xlabel('K'); axes[1].set_ylabel('Silhouette Score')
axes[1].set_title('Silhouette Analysis')
axes[1].grid(True, alpha=0.3)

best_k = list(K_range)[np.argmax(silhouettes)]
axes[1].axvline(best_k, color='g', linestyle='--', label=f'Best K={best_k}')
axes[1].legend()
plt.tight_layout()
plt.show()
print(f'Best K by Silhouette: {best_k}')
```

## Key Takeaways
- **Silhouette score** is the most reliable automated method
- **Elbow method** is quick and visual but subjective
- Always combine quantitative methods with **domain knowledge**
- Consider **GMM + BIC** as an alternative that automatically determines the number of clusters
