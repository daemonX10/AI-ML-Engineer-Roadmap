## Question
Explain Logistic Regression in depth. Why is it called regression when it's classification?

## Detailed Answer

### The Model
Logistic Regression models the probability of the positive class:
$$P(y=1|x) = \sigma(w^T x + b) = \frac{1}{1 + e^{-(w^T x + b)}}$$

### Why "Regression"?
It **regresses** the log-odds (logit) as a linear function of features:
$$\log\frac{P(y=1)}{P(y=0)} = w^T x + b$$
The log-odds is continuous and unbounded — it's a regression on this transformed space.

### Loss Function (Binary Cross-Entropy)
$$\mathcal{L} = -\frac{1}{N}\sum[y_i \log(\hat{y}_i) + (1-y_i)\log(1-\hat{y}_i)]$$
- No closed-form solution → optimize with gradient descent or Newton's method
- Convex → guaranteed global optimum

### Key Properties
- **Linear decision boundary**: $w^T x + b = 0$ is a hyperplane
- **Coefficients are interpretable**: $e^{w_j}$ = odds ratio for feature $j$
- **Probabilistic output**: Directly outputs well-calibrated probabilities
- **Regularization**: L1 (Lasso-like) or L2 (Ridge-like) via C parameter

### Multi-class Extension (Softmax Regression)
$$P(y=k|x) = \frac{e^{w_k^T x}}{\sum_j e^{w_j^T x}}$$

### Assumptions
- Linear relationship between features and log-odds
- No multicollinearity (features not highly correlated)
- Independent observations

### When to Use
- Baseline for any classification task
- When interpretability matters
- When calibrated probabilities are needed
- High-dimensional sparse data (text classification with L1)
