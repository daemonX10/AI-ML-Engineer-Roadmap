## Question
Why Naive Bayes is called Naive?

## Detailed Answer

### Bayes' Theorem
$$P(y|x_1,...,x_n) = \frac{P(x_1,...,x_n|y) \cdot P(y)}{P(x_1,...,x_n)}$$

### The "Naive" Assumption
Naive Bayes assumes all features are **conditionally independent** given the class:

$$P(x_1,...,x_n|y) = \prod_{i=1}^{n} P(x_i|y)$$

This is called "naive" because this assumption is **almost never true** in real data. Features are usually correlated.

**Example**: In spam detection, features like "free" and "winner" are correlated (both appear in spam), but Naive Bayes treats them as independent.

### Why It Works Anyway
1. **Classification doesn't need correct probabilities** — only the correct ranking (which class has highest P)
2. **Dependencies often cancel out** when comparing classes
3. **Very few parameters** to estimate → low variance, works with small datasets
4. **Fast** to train (O(n·d)) and predict (O(d))

### Variants
| Variant | Feature Type | Distribution |
|---------|-------------|-------------|
| Gaussian NB | Continuous | $P(x_i|y) \sim \mathcal{N}(\mu_{iy}, \sigma_{iy}^2)$ |
| Multinomial NB | Counts (text) | Multinomial distribution |
| Bernoulli NB | Binary | Bernoulli distribution |

### When to Use
- Text classification (spam, sentiment) — Multinomial NB is a strong baseline
- Very small datasets
- Real-time classification (very fast inference)
- When features are actually approximately independent

## Key Takeaways
- Understanding the underlying principles is more important than memorizing formulas
- Always consider practical implications alongside theoretical knowledge
- Be ready to discuss trade-offs and when to apply each approach
