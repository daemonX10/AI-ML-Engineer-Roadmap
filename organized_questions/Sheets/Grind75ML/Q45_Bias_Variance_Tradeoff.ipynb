## Question
What is the Bias-Variance tradeoff?

## Detailed Answer

### Error Decomposition
For any model, the expected prediction error decomposes as:

$$\text{Error} = \text{Bias}^2 + \text{Variance} + \text{Irreducible Noise}$$

### Bias
Error from wrong assumptions in the model. High bias → **underfitting**.
- Model is too simple to capture the true relationship
- Linear model on non-linear data
- Consistently wrong predictions

### Variance
Error from sensitivity to training data fluctuations. High variance → **overfitting**.
- Model memorizes training data noise
- Deep tree, high-degree polynomial
- Predictions vary wildly across different training sets

### The Tradeoff
- Simple models: high bias, low variance
- Complex models: low bias, high variance
- Goal: find complexity level where total error is minimized

### Model Complexity Spectrum
```
Simple ←──────────────────────→ Complex
High Bias                        Low Bias
Low Variance                     High Variance
Underfitting                     Overfitting

Linear Reg → k-NN(large k) → DT → RF → Deep NN
```

### Reducing Each
| To Reduce Bias | To Reduce Variance |
|---------------|-------------------|
| More complex model | Regularization (L1/L2) |
| Add features | More training data |
| Decrease regularization | Dropout, early stopping |
| Use ensemble (boosting) | Use ensemble (bagging) |

## Key Takeaways
- Understanding the underlying principles is more important than memorizing formulas

- Be ready to discuss trade-offs and when to apply each approach
