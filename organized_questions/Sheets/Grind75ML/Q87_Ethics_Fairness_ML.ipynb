## Question
How do you ensure fairness and address bias in ML models?

## Detailed Answer

### Types of Bias
| Type | Source | Example |
|------|--------|--------|
| **Historical bias** | Real-world inequality in data | Gender pay gap in salary prediction |
| **Representation bias** | Dataset doesn't represent population | Facial recognition trained mostly on light skin |
| **Measurement bias** | Proxy variables correlate with protected attributes | Zip code as proxy for race |
| **Aggregation bias** | One model for all subgroups | Different disease presentation across demographics |
| **Evaluation bias** | Benchmark doesn't represent deployment | Testing on one demographic only |

### Fairness Metrics
- **Demographic parity**: $P(\hat{y}=1|A=0) = P(\hat{y}=1|A=1)$
- **Equalized odds**: Equal TPR and FPR across groups
- **Calibration**: $P(y=1|\hat{y}=p, A=a) = p$ for all groups
- **Individual fairness**: Similar inputs → similar outputs

### Mitigation Strategies

**Pre-processing:**
- Balanced dataset collection
- Reweighting or resampling underrepresented groups
- Removing or transforming proxy features

**In-processing:**
- Adversarial debiasing: Train adversary to predict protected attribute; penalize if successful
- Fairness constraints in objective function

**Post-processing:**
- Adjust thresholds per group to equalize rates
- Calibrate probabilities per group

### Best Practices
- Audit model across all demographic groups
- Document model cards (intended use, limitations, fairness metrics)
- Regular monitoring in production for fairness degradation

## Key Takeaways
- Understand the core concepts and trade-offs
- Be ready to discuss practical applications and alternatives
