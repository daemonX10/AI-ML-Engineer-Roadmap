# Q57: Underfitting vs Overfitting
**Topic:** Classical-ml | **Difficulty:** Easy | **Time:** 5 min
**Sheet:** Grind75ML

---

## Question
What is underfitting and how to avoid it? What is overfitting and how to avoid it?

## Detailed Answer

### Underfitting (High Bias)
Model is **too simple** to capture the underlying pattern.

**Symptoms**: Low training acc, low validation acc (both bad)

**Fixes**:
- Increase model complexity (more layers, more features)
- Train longer (more epochs)
- Better features / feature engineering
- Reduce regularization
- Use a more powerful model

### Overfitting (High Variance)
Model **memorizes training data** including noise.

**Symptoms**: High training acc, low validation acc (big gap)

**Fixes**:
- More training data (or data augmentation)
- Regularization (L1, L2, dropout)
- Early stopping
- Simpler model
- Ensemble methods
- Cross-validation for model selection

### The Ideal Model
```
Training Error:    Low
Validation Error:  Low
Gap:               Small
```

### Summary
| Aspect | Underfitting | Overfitting |
|--------|-------------|------------|
| Bias | High | Low |
| Variance | Low | High |
| Train error | High | Low |
| Val error | High | High |
| Fix | More complex model | Regularize/simplify |

## Key Takeaways
- Watch the **train-val gap** — it tells you which regime you're in
- If both errors are high → underfitting → increase capacity
- If gap is large → overfitting → regularize or get more data
- Use **learning curves** (error vs. data size) to diagnose
