## Question
What does an end-to-end ML pipeline look like in production?

## Detailed Answer

### End-to-End Pipeline
```
Data Ingestion → Feature Engineering → Feature Store → Training → Evaluation → 
Model Registry → Deployment → Monitoring → Retraining (loop)
```

### Key Components

**1. Data Pipeline**
- Ingestion: Kafka, Spark, Airflow
- Validation: Great Expectations, data schema checks
- Versioning: DVC, Delta Lake

**2. Feature Store**
- Centralized repository for features (Feast, Tecton)
- **Offline**: Batch features for training (high throughput)
- **Online**: Low-latency features for serving (Redis, DynamoDB)
- Prevents train-serve skew (same features in training and inference)

**3. Training**
- Experiment tracking: MLflow, W&B
- Compute: Kubernetes, SageMaker, GCP Vertex AI
- Hyperparameter tuning: Optuna, Ray Tune

**4. Model Registry**
- Version models with metadata
- Stage management: staging → production → archived
- MLflow Model Registry, SageMaker Model Registry

**5. Serving**
- REST API: Flask, FastAPI, TorchServe
- Batch inference: Spark, offline scoring
- Real-time: gRPC, Triton Inference Server

**6. Monitoring**
- Model performance: accuracy decay, prediction distribution
- Data drift: feature distribution changes
- System health: latency, throughput, error rates
