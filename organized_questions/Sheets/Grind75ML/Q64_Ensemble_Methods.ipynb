## Question
What are the types of ensemble methods and when to use each?

## Detailed Answer

### 1. Bagging (Bootstrap Aggregating)
Train multiple models on **random subsets** of data, aggregate predictions.
- Reduces **variance**
- Models trained in **parallel** (independent)
- Example: **Random Forest** (bagged decision trees + random features)
- Aggregation: majority vote (classification) or mean (regression)

### 2. Boosting
Train models **sequentially**, each focusing on previous model's errors.
- Reduces **bias** (and some variance)
- Models trained **sequentially** (dependent)
- Algorithms: **AdaBoost, XGBoost, LightGBM, CatBoost**
- Can overfit if not regularized

### 3. Stacking
Train a **meta-model** on the predictions of base models.
- Base models (diverse): RF, XGBoost, SVM, NN
- Meta-model (simple): Logistic regression or linear model
- Uses cross-validation to prevent meta-model from overfitting to base model predictions

### 4. Voting/Averaging
Combine predictions from different models.
- **Hard voting**: Majority vote
- **Soft voting**: Average probabilities (usually better)
- **Weighted**: Assign higher weight to better individual models

### When to Use
| Method | Use When |
|--------|----------|
| Bagging (RF) | High variance, want stability |
| Boosting (XGB) | Want best accuracy on tabular data |
| Stacking | Competition/production, diverse base models |
| Voting | Quick improvement from existing models |

## Key Takeaways
- **XGBoost/LightGBM** (boosting) is the default for tabular data competitions
- **Random Forest** (bagging) is the low-tuning, robust option
- **Stacking** gives the best results but is more complex to implement
- Ensemble diversity is key — use different model types for best results
