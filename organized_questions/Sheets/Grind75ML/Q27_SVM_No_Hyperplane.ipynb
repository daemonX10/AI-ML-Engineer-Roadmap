## Question
What happens when there is no clear Hyperplane in SVM?

## Detailed Answer

When data is **not linearly separable**, a perfect hyperplane doesn't exist. SVM handles this in two ways:

### 1. Soft Margin SVM (C-SVM)
Allow some misclassifications by introducing **slack variables** $\xi_i$:

$$\min_{w,b,\xi} \frac{1}{2}\|w\|^2 + C\sum_{i=1}^{n}\xi_i$$
$$\text{s.t. } y_i(w \cdot x_i + b) \geq 1 - \xi_i, \quad \xi_i \geq 0$$

- **C** controls the trade-off: High C → few misclassifications (overfit), Low C → wider margin (underfit)
- $\xi_i = 0$: correctly classified beyond margin
- $0 < \xi_i < 1$: correctly classified but within margin
- $\xi_i \geq 1$: misclassified

### 2. Kernel Trick
Map data to a **higher-dimensional space** where it becomes linearly separable:

$$K(x_i, x_j) = \phi(x_i) \cdot \phi(x_j)$$

Common kernels:
| Kernel | Formula | Best For |
|--------|---------|----------|
| Linear | $x^T x'$ | Already separable data |
| RBF (Gaussian) | $\exp(-\gamma\|x-x'\|^2)$ | Most cases (default) |
| Polynomial | $(\gamma x^T x' + r)^d$ | Interaction features |

The kernel trick computes the dot product in high-d space **without explicitly transforming** — this is the key efficiency.

### In Practice: Combine Both
Use soft margin + kernel: allows flexibility in both feature space and misclassification tolerance.

## Key Takeaways
- **Soft margin** allows controlled misclassification via C parameter
- **Kernel trick** projects to higher dimensions where linear separation exists
- **RBF kernel** is the go-to default when data isn't linearly separable
- Tune **C** and **gamma** together using grid search or Bayesian optimization
