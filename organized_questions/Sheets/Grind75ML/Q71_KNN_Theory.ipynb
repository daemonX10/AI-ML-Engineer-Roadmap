## Question
Explain KNN. What is the impact of K? How do you choose the distance metric?

## Detailed Answer

### How KNN Works
1. Store all training data (lazy learner — no training phase)
2. For a new point: find K closest training points
3. Classification: majority vote; Regression: average

### Impact of K
- **K=1**: Overfitting — decision boundary is very complex (every point)
- **Large K**: Underfitting — smooth boundary, approaches majority class
- **Odd K** for binary classification (avoids ties)
- Typically: K = 3, 5, or ~$\sqrt{N}$

### Distance Metrics
| Metric | Formula | Best For |
|--------|---------|----------|
| **Euclidean** ($L_2$) | $\sqrt{\sum(x_i-y_i)^2}$ | Continuous, same scale |
| **Manhattan** ($L_1$) | $\sum|x_i-y_i|$ | High-d, sparse data |
| **Cosine** | $1 - \frac{x \cdot y}{\|x\|\|y\|}$ | Text/embeddings |
| **Hamming** | Fraction of positions that differ | Categorical/binary |
| **Minkowski** ($L_p$) | $(\sum|x_i-y_i|^p)^{1/p}$ | General (p=1→Manhattan, p=2→Euclidean) |

### Practical Considerations
- **Feature scaling is CRITICAL** — unscaled features with large ranges dominate
- Curse of dimensionality: KNN degrades badly in high dimensions
- Computational: O(N·d) per prediction; use KD-tree or Ball-tree for speedup
- Not suitable for real-time prediction on large datasets

### Weighted KNN
Weight neighbors by inverse distance — closer neighbors have more influence.
Reduces sensitivity to K choice.
