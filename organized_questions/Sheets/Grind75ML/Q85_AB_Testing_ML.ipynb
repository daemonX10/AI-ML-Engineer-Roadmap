## Question
How would you A/B test a new ML model in production?

## Detailed Answer

### The Setup
- **Control (A)**: Current production model
- **Treatment (B)**: New candidate model
- Traffic split: Typically 50/50 or 90/10 (start cautious)

### Steps
1. **Define metrics**: Primary (revenue, CTR) + guardrail (latency, error rate)
2. **Calculate sample size**: Based on desired statistical power, MDE, baseline rate
3. **Random assignment**: Hash user/session ID for consistent routing
4. **Run experiment**: Collect data for sufficient duration (≥1-2 weeks for seasonality)
5. **Analyze**: Statistical significance test (t-test, proportions test)
6. **Decision**: Ship if significant improvement + no guardrail violations

### ML-Specific Considerations
- **Novelty effect**: Users may interact differently with new recommendations initially
- **Position bias**: Reranking changes click patterns regardless of quality
- **Feedback loops**: Model A and B may see different training data over time
- **Feature store consistency**: Both models should use same feature values

### Alternatives to A/B Testing
| Method | When |
|--------|------|
| **Shadow mode** | Run B in parallel, compare offline (low risk) |
| **Interleaving** | Mix results from A and B in same response (faster convergence) |
| **Multi-armed bandit** | Dynamically allocate more traffic to better variant |
| **Canary deploy** | Route 1-5% traffic first, monitor, then increase |
