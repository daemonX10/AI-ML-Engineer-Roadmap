## Question
Compare PCA, t-SNE, and UMAP for dimensionality reduction.

## Detailed Answer

### PCA (Principal Component Analysis)
- **Linear** method — finds orthogonal directions of maximum variance
- Preserves **global structure** (large distances)
- Can reconstruct data (inverse transform)
- Interpretable components (loadings)
- Fast: O(min(nd², n²d))
- Use for: Pre-processing, feature reduction, compression

### t-SNE (t-distributed Stochastic Neighbor Embedding)
- **Non-linear** — preserves local neighborhood structure
- Converts distances to probabilities; minimizes KL divergence between high-d and low-d distributions
- Excellent for **visualization** (2D/3D)
- Slow: O(n²) or O(n·log(n)) with approximations
- **Not suitable for pre-processing** (non-parametric, can't transform new points)
- Perplexity is the key hyperparameter (~5-50)

### UMAP (Uniform Manifold Approximation and Projection)
- **Non-linear** — based on topological data analysis
- Preserves **both local and global** structure better than t-SNE
- Much faster than t-SNE: O(n·log(n))
- **Can transform new points** (has a `transform()` method)
- Key parameters: n_neighbors (local/global), min_dist (cluster tightness)

### Comparison
| Feature | PCA | t-SNE | UMAP |
|---------|-----|-------|------|
| Linear/Non-linear | Linear | Non-linear | Non-linear |
| Structure preserved | Global | Local | Both |
| Speed | Fast | Slow | Medium |
| Transform new data | Yes | No | Yes |
| Use case | Preprocessing | Visualization | Visualization + preprocessing |
| # components | Any k ≤ d | 2-3 | Any (usually 2-3) |
