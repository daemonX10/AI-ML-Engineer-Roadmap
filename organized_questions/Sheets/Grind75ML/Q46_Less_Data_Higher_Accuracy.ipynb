## Question
How can less Training Data give Higher Accuracy?

## Detailed Answer

Counterintuitive but possible in several scenarios:

### 1. Removing Noisy / Mislabeled Data
If training data contains errors, removing bad samples improves learning.
- Data cleaning → higher quality → better model
- Label noise can hurt more than missing data

### 2. Removing Outliers
Outliers can bias model parameters significantly.
- KNN: outliers directly affect predictions
- Linear models: outliers pull the regression line

### 3. Reducing Class Imbalance
Undersampling the majority class can improve minority class performance.
- Overall accuracy might drop, but F1/recall on minority class improves
- Important for anomaly detection, medical diagnosis

### 4. Removing Redundant Features
Fewer features = less noise, faster convergence, better generalization.
- Feature selection can dramatically improve performance
- Curse of dimensionality: more features can hurt with limited samples

### 5. Curriculum Learning
Training on easy examples first, then gradually adding harder ones.
- At any point, using a subset of data can give better intermediate performance

### 6. Active Learning
Selecting the most informative samples gives higher accuracy per sample.
- 1000 actively selected samples > 5000 random samples

### 7. Transfer Learning
A pre-trained model fine-tuned on 100 labeled examples can outperform a model trained from scratch on 10,000 examples.

## Key Takeaways
- Understanding the underlying principles is more important than memorizing formulas
- Always consider practical implications alongside theoretical knowledge
- Be ready to discuss trade-offs and when to apply each approach
